## Основные теоретические положения.

#### Бинарный метод.

Бинарный(двоичный) метод возведения в степень — это метод, позволяющий возводить любое число в степень $n\in N$ за O(log n) умножений (вместо n умножений при обычном подходе). В этом методе мы проходим по степени n в двоичном виде слева направо, начинаем с самого левого бита, и при каждом переходе на следующий бит возводим число в квадрат, если бит, на который мы перешли, равен 1, то после возведения в квадрат, еще и умножаем на исходное число. В итоге, с помощью этого метода мы возвели число в n-ю степень за $\lambda (n)+\nu (n)-1$ умножений($\lambda (n) = \lceil log_2(n) \rceil$, $\nu (n)$ - количество единиц в двоичной записи числа).

#### Метод множителей.

Метод множителей - метод быстрого возведения в степень, путем разложения степени $n\in N$ на 2 множителя. После разложения происходит возведение числа в степень, равную меньшему множителю. Затем итоговое число возводится в степень, равную второму множителю. Если степень n - простое число, то сначала с помошью этого метода возводим число в степень n-1, а потом домножаем на исходное число.

#### Аддитивная цепочка.

Аддитивная цепочка — последовательность натуральных чисел, начинающаяся с единицы, в которой каждый последующий элемент является суммой каких-то двух предшествующих элементов (в том числе, возможно использование одного и того же предшествующего элемента — удвоение). Формально, в аддитивной последовательности ${\displaystyle a{i}}$ выполнены условия:
* ${\displaystyle a_{0}=1}$
* для любого ${\displaystyle i>0}$, ${\displaystyle a_{i}=a_{j}+a_{k}}$, где ${\displaystyle j,k<i}$

Звездной аддитивной цепочкой называется такая аддитивная цепочка, что $a_{i} = a_{i-1}+a_k$.

#### Алгоритм Яо для вычисления приближенных аддитивных цепочек.

Цепочка Яо - аддитивная цепочка, состаящая из 2 аддитивных цепочек, первая цепочка равна $Y_k(n):1,2,4,\ ...\, 2^{\lambda (n)}$, вторая аддитивная цепочка составляется на основе первой. $d(z)$ - сумму всех разрядов, цифра которых равняется z. Алгоритм Яо составляет для числа $n\in N$ цепочку Яо. Сначала алгоритму подается число $n\in N$ и $k\in N$. Далее алгоритм разбивает число на сумму: $\sum_{i=0}^{2^k-1}{2^{ik}a_i}$, где $a_i$ - цифра i-го разряда числа. Потом алгоритм считает $zd(z)$ для всех чисел от 1 до $2^k-1$. $zd(z)$ получается с помощью сложения чисел из $Y_k(n)$. Наконец сложением всех $d(z)$ получаем нужное нам число n. Все промежуточные значения записывются во вторую аддитивную цепочку цепочки Яо.

#### Алгоритм дробления вектора индексов.

Вектор индексов - последовательность натуральных числе длиной m-1, по которой можно построить аддитивную звездную цепочку длиной m. Каждый элемент в векторе - это индекс элемента звездной цепочки, который нужно прибавить к i-му элементу, чтобы получить i+1 элемент. Данный алгоритм нужен для того, чтобы найти для заданного числа $n\in N$ минимальную звездную цепочку. Алгоритм сначала с помощью внешнего цикла перебирает все длины m вектора индексов, где $\underline{l}(n) \le m \le \overline{l}(n)$. $\underline{l}(n) = \lceil lbn \rceil$, $\overline{l}(n) = \lambda(n) + \nu(n)-1$. Далее происходит деление вектора индексов на 2 части: $\{r_i\}_{i=1}^q$ и $\{\rho\}_{j=q+1}^m$, где $1\le q\le m-1$. Далее с помощью внутреннего цикла происходит перебор первой части вектора, в нем считается максимальное $a_{max}$ и минимальное значение $a_{min}$ итогового числа. Если $a_m \in [a_{min};a_{max}]$ , то с помощью второго внутреннего цикла происходит перебор всех 2 частей вектора, в ином случае алгоритм переходит на следующий вектор. Если $a_m = n$, то алгоритм завершает свою работу.

#### Гипотеза Шольца-Брауэра

Гипотеза Шольца-Брауэра для звездной цепочки заключается в $l^*(n^2-1)\le l^*(n)+n-1$, где $l^*(n)$ - длина минимальной аддитивной цепочки для числа n.

####  Бинарный метод и метод множителей:


```python
def BinaryMethod(a, n):
    power = bin(n)
    power = power[2::]
    m_Power = power
    counter = 0
    m_A = a

    for i in range(1, len(m_Power)):
        if power[i] == '1':
            m_A *= m_A
            m_A *= a
            counter += 2
        else:
            m_A *= m_A
            counter += 1
    return (m_A, counter)


def MyFact(n):
    i = 2
    for i in range(i,n):
        if n % i == 0:
            return i
    return n


def MM(a, power):
    if power == 0:
        return (1, 0)
    elif power == 1:
        return (a, 0)

    m_Fact = MyFact(power)
    if m_Fact == power:
        answer = MM(a, power - 1)
        return (answer[0] * a, answer[1] + 1)

    temp = power // m_Fact

    ans, count1 = BinaryMethod(a, m_Fact)
    ans1, count2 = BinaryMethod(ans, temp)
    countOperations = count1 + count2

    return ans1, countOperations


if __name__ == '__main__':
    
    x = 1
    n = [45, 87, 130481]
    
    for i in n:
        print("\n")
        print("Степень возведения x: ", i)
        print("Бинарный метод:", BinaryMethod(x, i))
        print("Метод множителей:", MM(x, i))
        print("\n")
```

#### Алгоритм Яо:


```python
import math

def Fdz(z, dz):
    result_Fdz = [dz]
    m_Z = dz
    bin_z = bin(z)[3:]
    for i in bin_z:
        if i == '0':
            m_Z += m_Z
            result_Fdz.append(m_Z)
        elif i == '1':
            m_Z += m_Z
            result_Fdz.append(m_Z)
            m_Z += dz
            result_Fdz.append(m_Z)
    return result_Fdz


def Yao(x, k):
    m_Chain = []
    m_X = x
    while (m_X != 0):
        m_Chain.append(m_X % 2 ** k)
        m_X //= 2 ** k
    chain = []
    for i in range(int(math.log(x, 2)) + 1):
        chain.append(2 ** i)
    count = 0
    for z in range(1, 2 ** (k) + 1):
        dz = 0
        for i in range(0, len(m_Chain)):
            if m_Chain[i] == z:
                dz += 2 ** (k * i)
                if dz not in chain:
                    chain.append(dz)
        if (dz == 0):
            continue
        m_Dz = Fdz(z, dz)
        for indexDz in m_Dz:
            if indexDz not in chain:
                chain.append(indexDz)
        if len(m_Dz) == 0:
            continue
        count += m_Dz[-1]
        if count not in chain:
            chain.append(count)
    return chain



if __name__ == "__main__":
    
    m = 2  
    n = [8,20,70,105,140]
    
    print('\n')
        
    for i in n:
        result0 = Yao(i, m)
        result1 = Yao(i, m + 1)
        result2 = Yao(i, m + 2)
        result3 = Yao(i, m + 3)

        print("n = ", i, '\n')
        print("k = ", m, ", АЦ:", result0, ", Длина АЦ: ", len(result0))
        print("k = ", m + 1, ", АЦ:", result1, ", Длина АЦ: ", len(result1))
        print("k = ", m + 2, ", АЦ:", result2, ", Длина АЦ: ", len(result2))
        print("k = ", m + 3, ", АЦ:", result3, ", Длина АЦ: ", len(result3))
        print('\n')
```

#### Алгоритм дробления вектора индексов

```python
import time
import math


def Find_Ind_Vector(vector, index_def = 1):
    max_Index = index_def + len(vector) - 1
    for i in range(1, len(vector) + 1):
        vector[-i] += 1
        if vector[-i] <= max_Index:
            break
        if i == len(vector):
            vector.append(1)
            break
        vector[-i] = 1
        max_Index -= 1


def New_Chain(r, ro):
    q = len(r)
    chain = [1]
    for index in r:
        chain.append(chain[-1] + chain[index - 1])
    for index in ro:
        chain.append(chain[-1] + chain[index - 1])
    return chain


def Split_Index_Vector(n):
    m = int(log(n, 2))
    up = m + bin(n).count('1') - 1
    while m <= up:
        q = int(m / 2)
        if q == 0:
            q = 1
        r = [1 for i in range(q)]
        while len(r) == q:
            ro = [1 for i in range(m - q)]
            chain = New_Chain(r, ro)
            if chain[-1] == n:
                return chain
            a_Max = chain[q] * 2 ** (m - q)
            a_Min = chain[q] + m - q
            if n < a_Min or n > a_Max:
                Find_Ind_Vector(r)
                continue
            while len(ro) == m - q:
                chain = New_Chain(r, ro)
                if chain[-1] == n:
                    return chain
                Find_Ind_Vector(ro, q + 1)
            Find_Ind_Vector(r)
        m += 1
    return chain



if __name__ == "__main__":
    
    n = int(input("Введите n: "))
    
    start = time.time()
    chain = Split_Index_Vector(n)
    finish = time.time()
    
    print("Аддитивная цепочка:", chain)
    print("Длина аддитивной цепочки:", len(chain))
    print("Время работы: ", float(finish - start))
```

#### Проверка первого следствия из теоремы Брауэра

```python


import pylab as plt

def Summ_1(ylnxi, Len, n, lnxi):
    return n * sum(ylnxi) - sum(Len) * sum(lnxi)

def Summ_2(n, lnxi, lnx2):
    return n * sum(lnx2) - sum(lnxi) * sum(lnxi)

def Summ_3(Len, lnxi, b):
    return sum(Len) - b * sum(lnxi)




if __name__ == "__main__":
    
    n = 100
    x = [i for i in range(1, n)]
    Len = [len(Split_Index_Vector(i)) for i in range(1, n)]
    lnxi = [math.log(i, math.e) for i in range(1, n)]
    lnx2 = [pow(math.log(i, math.e),2) for i in range(1, n)]
    ylnxi = [Len[i-1] * lnxi[i-1] for i in range(1, n)]

    b = Summ_1(ylnxi, Len, n, lnxi) / Summ_2(n, lnxi, lnx2)
    a = Summ_3(Len, lnxi, b) / (n - 1)

    y = [a + b*lnxi[i-1] for i in range(1, n)]

    plt.figure(figsize=(10, 8))
    plt.scatter(x, Len, color = 'red', marker = '.')
    plt.plot(x, y)
    plt.show()

```

#### Вывод:

Аппроксимирующая прямая $l(n)$ схожа с графиком логарифма. Это подтвержает следствие теоремы Брауэра, поскольку в этом случае на бесконечности $\lambda(n)$ и $l(n)$ будут эквивалентны.
