## Основные теоретические положения.
Пусть, $G$ - группа с нейтральным элементом $e$. $x$ - множество. $G$ действует на $x$, если задана операция $G * X \to G$.

$( \forall x \in X$, $\forall g, h \in G )$ выполняются следующие свойства:
- $ex = x$
- $g(hx) = (gh)x$

Орбита $x \in X$ под действием $G$, называется множество, $Gx = \{ gX|g \in G \}$, количество элементов в орбите - длина орбиты.

Неподвижная точка эл-та $g \in G$ называется те $x \in X$ для которых выполняется $gx = x$.

### Th(Бёрнсайда)
Количество орбит, действия $G$ на $X$ равно $N = \frac{1}{\#G}\sum\limits_{g\in G}\#X^{g}$

## Реализация функции отображения раскраски по заданному номеру.

```python
sage: from sage.plot.plot3d.shapes import ColorCube

def translateNumbersToColors(numbers):
    colors = []
    colorsGlossary = {1 : 'red', 2 : 'green', 3 : 'blue'}
    for number in numbers:
        colors.append(colorsGlossary[number])
    return colors
        
def drawCube(colors):
    cube = ColorCube([1, 1, 1], translateNumbersToColors(colors))
    cube.show()
    
drawCube([3, 2, 1, 1, 2, 3])
```

```python
def rotate90_1(cubeData):
    rotatedCube = []
    transformsGlossary = {1 : 1, 2 : 3, 3 : 5, 4 : 4, 5 : 6, 6 : 2}
    for index in range(1, 7):
        rotatedCube.append(cubeData[transformsGlossary[index] - 1])
    return rotatedCube

drawCube(rotate90_1([3, 2, 1, 1, 2, 3]))
```
