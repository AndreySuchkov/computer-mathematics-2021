## Основные теоретические положения

### Определение 1
$\sqsupset G$ - группа с нейтральным элементом $e$, а $X$ - множество. $G$ действует на $X$, если задана операция $G \times X \to X$, обладающая для $\forall x \in X$ и $\forall g,h \in G$ следующими свойствами: 

1) $ex = x$

2) $g(hx) = (gh)x$

где $gx$ - это образ пары $(g,x)$.

###  Определение 2
$\sqsupset$ $X$ - множество. Множество всех биективных функций с $X \to X$ с операцией композиции называется симметрической группой на множестве $X$ и обозначается как $S_{X}$

Важно отметить, что любой гомоморфизм $\varphi: G \to S_{X}$ определяет действие группы $G$ на $X$ по правилу:
        
$gx = \varphi(g)(x)$

### Определение 3
Орбитой элемента $x \in X$ под действием $G$ называется множество $Gx = \{gx | g \in G\}$. Количество элементов в данной орбите называется длиной орбиты. Любые две орбиты либо не пересекаются, либо совпадают. Таким образом, множество $X$ разбивается в дизъюнктивное объединение орбит.

### Определение 4
Неподвижными точками элемента $g \in G$ называются те $x \in X$, для которых $gx = x$. Множество неподвижных точек элемента g обозначается через $X^{g}$.


### Теорема (лемма Бернсайда)
Количество орбит действия группы $G$ на множестве $X$ равно:

$n = \frac{1}{\#G}\sum\limits_{g\in G}$

### Замечание 
Лемма Бернсайда будет непосредственно использована в задаче подсчета числа раскрасок с точностью до поворотов. Те раскраски, которые находятся в одной орбите, можно получить одну из другой, а значит их нужно посчитать всего один раз. Поэтому нас будет интересовать именно общее число орбит, а не раскрасок.

### Определение 5
Пусть $I$ - произвольное множество, а $C$ - множество цветов. Раскраской множества $I$ называется функция из $I$ в $C$ (каждому элементу множества ставится в соответствие некоторый цвет). Множество всех раскрасок обозначается через $C^{I}$. 

### Лемма 1
Пусть $G$ группа, действующая на множестве $I$. Тогда формула $gf(i) = f(g^{-1}i)$, где $g \in G$, $i \in I$, а $f \in C^{I}$ задает действие $G$ на $C^{I}$.

### Лемма 2
Для элемента $g \in G$ и функции $f \in C^{I}$ следующие условия эквивалентны:

1. $gf = f$

2. $f(i) = f(g^{n}i)$, $\forall i \in I $ и $n \in \mathbb{Z}$

3. Если $i,j \in I$ лежат в одном цикле в циклической записи перестановки $\varphi(g)$, то $f(i) = f(j)$ 

### Следствие
Количество раскрасок из $C^{I}$, которые сохраняет данный элемент $g \in G$, равно $\#X^g=\#C^{c(\varphi(g))}$, где $c(\varphi(g))$ - количество независимых циклов перестановки $\varphi(g)$ (считая циклы длины 1).

### Задача
Сколькими способами можно раскрасить грани тетраэдра используя краски 4 цветов (с точностью до поворотов тетраэдра).

### Решение
Множество $I={1,2,3,4}$ состоит из 4 элементов - граней тетраэдра.
Множество $C={a,b,c,d}$ состоит из 4 цветов.
Обозначим за $X=C^I$ множество всех раскрасок тетраэдра. 
Всего существует $|C|^{|I|}=4^4$ возможных раскрасок тетраэдра, но некоторые из них будут совмещаться при поворотах.


Все повороты тетраэдра образуют группу $G$, изоморфную знакопеременной подгруппе $A_4$ симметрической группы $S_4$. Это значит, что все повороты тетраэдра можно записать как четные подстановки множества из 4 элементов (граней).
$G = \{id, (123), (132), (124), (142), (134), (143), (234), (243), (12)(34), (13)(24), (14)(23)\}$ 

По сути, это четные перестановки множества граней $I={1,2,3,4}$, которые соответствуют поворотам тетраэдра. Среди этих перестановок есть одна тождественная, 8 циклов длины 3 и 3 перестановки-произведения пары независимых транспозиций.

Группа $G$ действует на множестве раскрасок $X$ в соответствии с Леммой 1. Чтобы найти число раскрасок с точностью до поворотов тетраэдра, найдем число орбит этого действия $G$ на $X$. Для этого воспользуемся Леммой Бернсайда:

$n = \frac{1}{\#G}\sum\limits_{g\in G} \#X^{g}$


Для каждого $g\in G$ необходимо найти число неподвижных элементов множества $\#X^g$, то есть число раскрасок, которые переходят в себя при действии элементом $g$. 
Чтобы вычислить эти значения, применим Следствие из Леммы 2 для каждого элемента группы $G$


$\#X^{id} = 4^{4}= 256$   все раскраски будут неподвижными для тождественного движения.

$\#X^{z} = 4^{2}= 16$,   где $z$ это один из 8 циклов длины 3, который представляется в виде произведения двух независимых циклов длины 3 и 1. Например, $z=(123)(4)$.

$\#X^{t_2} = 4^{2}= 16$, где $t_2$ это одна из 3 перестановок, состоящих из двух независимых транспозиций, то есть произведения двух независимых циклов длины 2. Например, $t_2 = (12)(34)$.

Кроме того, нам потребуется число элементов в группе $G$

$\#G = \#A_4= 12$


Наконец, по лемме Бернсайда число орбит (число различных раскрасок) равно:

$n = \frac{1}{12}(256 + 8\cdot 16 + 3\cdot 16 ) = \frac{256 + 11\cdot 16}{12} = \frac{432}{12} = 36$

### Реализация функции отображения раскраски по заданному номеру

Сначала случайно раскрасим 4 грани тетраэдра в 4 цвета:

```python
import math
import random
from sage.plot.plot3d.shapes import Text

def PaintRandomColor():
    pallet = []
    for color in range(4):
        pallet.append(random.randint(1,4))
    return pallet

pallet = PaintRandomColor()
print(pallet)
```

```python
def BuildingTheTetrahedron(pallet):  
    m_Color = {1: 'red', 2: 'green', 3: 'blue', 4: 'purple'}
    T1 = polygon3d([(0,0,0), (math.sqrt(3)/2, 0.5, 0), (0,1,0)], color = m_Color[pallet[0]])
    T2 = polygon3d([(math.sqrt(3)/2, 0.5, 0), (0,1,0), (math.sqrt(3)/6, 0.5, math.sqrt(2)/math.sqrt(3))], color = m_Color[pallet[1]])
    T3 = polygon3d([(0,0,0), (0,1,0), (math.sqrt(3)/6, 0.5, math.sqrt(2)/math.sqrt(3))], color = m_Color[pallet[2]])
    T4 = polygon3d([(0,0,0), (math.sqrt(3)/2, 0.5, 0), (math.sqrt(3)/6, 0.5, math.sqrt(2)/math.sqrt(3))], color = m_Color[pallet[3]])
    m_Edge = Text("1", color = 'black').translate(0.3,0.5,-0.1) + Text("2", color = 'black').translate(0.5, 0.2, 0.3) + Text("3", color = 'black').translate(0.5, 0.7, 0.3) + Text("4", color = 'black').translate(0,0.5,0.5)
    return T1 + T2 + T3 + T4 + m_Edge
T = BuildingTheTetrahedron(pallet)
T.show()
```

#### Поворот тэтраэдара на 120$^{\circ}$ относительно первой грани и противоположной вершины:

```python
def Foo_1(pallet):
    m_Pallet = [pallet[0], pallet[3], pallet[1], pallet[2]]
    m_T = BuildingTheTetrahedron(m_Pallet)
    T.show()
    m_T.show()
Foo_1(pallet)
```

### Решение задачи в общем виде:

Нас интересует количество орбит, потому каждая орбита это множество раскрасок тетраэдра, которые можно получить друг из друга с помощью поворотов.

По Лемме Бёрнсайда это число находится:
$n = \frac{1}{\#G}\sum\limits_{g\in G} \#X^{g}$, где
$\#X^{g}$ это число раскрасок, которые при действии элемента g переходят в себя

Если $g = id$, то $\#X^g = \#C^{\#I} = M^4$

Если $g = (123) = (123)(4) \Rightarrow \#X^g = \#C^{c(g)} = M^2$ таких элементов $8$ 

Если $g = (12)(34)\Rightarrow \#X^g = \#C^{c(g)} = M^2$ таких элементов $3$ 

$n = \frac{1}{12} \cdot (M^4 + 11 \cdot M^2)$ число орбит, т.е. число различных раскрасок (с точностью до поворотов).
