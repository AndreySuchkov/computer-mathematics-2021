## Основные теоретические положения

##### Определение действия группы на множество.

Пусть $G$ - группа с нейтральным элементом $e$, а $X$ - множество. $G$ действует на $X$, если задана операция $G \times X \to X$, обладающая для $\forall x \in X$ и $\forall g,h \in G$ следующими свойствами:

1) $ex = x$

2) $g(hx) = (gh)x$

где $gx$ - это образ пары $(g,x)$.


##### Определение орбиты.

Орбитой элемента $x \in X$ под действием $G$ называется множество $Gx = \{gx | g \in G\}$. Количество элементов в данной орбите называется длиной орбиты. Любые две орбиты либо не пересекаются, либо совпадают. Таким образом, множество $X$ разбивается в дизъюнктивное объединение орбит.


##### Определение неподвижных точек.

Неподвижными точками элемента $g \in G$ называются те $x \in X$, для которых $gx = x$. Множество неподвижных точек элемента $g$ обозначается через $X^{g}$.


##### Лемма Бёрнсайда.

Количество орбит действия группы $G$ на множестве $X$ равно:
$\frac{1}{|G|}\sum\limits_{g\in G}|X^{g}|$

**Доказательство леммы.**

Обозначим число орбит через $N$. Каждый элемент $x\in X$ лежит в орбите $Gx$. Сопоставим ему число $\frac{1}{|Gx|}$. Сумма этих чисел по всем $x$ из данной орбиты $O$ равна 1. ($O$ раз складываем число $\frac{1}{|O|}$ с самим собой). Поэтому количесво орбит можно вычислить по формуле $N = \sum\limits_{x\in X}\frac{1}{|Gx|}$. Подлсавляя сюда формулу для длины орбиты $|Gx|\cdot |St(x)| = |G|$ получим $N = \sum\limits_{x\in X}\frac{|St(x)|}{|G|} = \frac{1}{|G|}\sum\limits_{x\in X}|St(x)|$. Используя формулу $\sum\limits_{x\in X}|St(x)| = \sum\limits_{g\in G}|X^g|$ получим $N = \frac{1}{|G|}\sum\limits_{g\in G}|X^g|$, что и требовалось доказать.


##### Определение множества раскрасок.

Пусть $I$ - произвольное множество, а $C$ - множество цветов.Расскраской множества $I$ называется функция из $I$ в $C$. Множество всех раскрасок обозначается через $C^I$. $|C^I| = |C|^{|I|}$


##### Лемма-Определение 1

Пусть $G$ группа, действующая на множестве $I$. Тогда формула $gf(i) = f(g^{-1}i)$, где $g \in G$, $i \in I$, а $f \in C^{I}$ задает действие $G$ на $C^{I}$.


**Лемма 2**

Для элемента $g \in G$ и функции $f \in C^{I}$ следующие условия эквивалентны:

1. $gf = f$
2. $f(i) = f(g^{n}i)$, $\forall i \in I $ и $n \in \mathbb{Z}$
3. Если $i,j \in I$ лежат в одном цикле в циклической записи перестановки $\varphi(g)$, то $f(i) = f(j)$


**Количество раскрасок.**

Количество раскрасок из $C^{I}$, которые сохраняет данный элемент $g \in G$ равно $|C|^{c(\varphi(g))}$, где $c(\varphi(g))$ - количество независимых циклов перестановки $\varphi(g)$ (считая циклы длины 1).


**Доказательство.**

Обозначим через $\langle g \rangle i$ орбиту элемента $i$ под действием циклической подгруппы, порождённой $g$ (упорядоченные элементы такой орбиты - это и есть цикл перестановки $\varphi(g)$). Пусть $Q$ - множество различных орбит. Если $gf=f$, то известно, что все элементы одной орбиты отображаются в одно и тоже под действием функции $f$. Поэтому $f$ однозначно определяется функцией $\tilde f : Q \rightarrow С$, где $\tilde f (\langle g \rangle i) = f(i)$. Другими словами, множество $(C^I)^g$ биективно отображается на множество $C^Q$, а значит эти множества имеют одинаковые количества элементов. Количесвто всех функций из $Q$ в $C$ равно $|C|^{|Q|}$, при этом $Q = |c(\varphi(g))|$. ЧТД.

### Задание:

Сколькими способами можно составить бусы из 161 бусинок 6 различных цветов (с точностью до поворотов и осевых симметрий)?

### Решение:

Итак, имеем упорядоченный набор $\{ x_{1}, x_{2}, x_{3}, ... , x_{161} \}$, где $x_{i} \in \{ a, b, c, d, e, f \}$. Т.е. 161 бусинка, каждая из которых может быть раскрашена в 6 различных цветов. Существует всего $6^{161}$ таких раскрасок, но не все из них соответствуют различным бусам.

По условию есть две операции, при применении которых будут получаться одинаковые бусы: поворот и симметричное отображение. Для них будем считать, что бусы до применения этих операций - такие же, как и после.

Таким образом на множестве всех раскрасок $X$ действует группа поворотов и осевых симметрий. Операция в группе - композиция, она является ассоциативной, нейтральный элемент - тождественное отображение, обратный элемент к осевой симметрии - сама эта симметрия, обратный элемент к повороту - поворот в другую сторону. Обозначим эту группу, как $G$.

Группа $G$ естественным образом действует на множестве мест $\{1, 2, 3, ..., 161\}$, т.е. задано отображение из $G$ в симметрическую группу $S_{161}$. 

Обозначим через $\tau$ поворот, переводящий каждое место в следующее(двигаем вправо), а последнее - в первое. Через $\sigma$ обозначим осевую симметрию относительно оси проходящую между 1 и 161, и 81.


Некоторые перестановки в циклической форме из $S_{161}$ в которые переходят элементы группы $G$:

$e = (1)(2)(3)...(161)$

$\tau = (1,2,3,...,161)$

$\tau^{161} = e = (1)(2)(3)...(161)$

$\sigma = (1,161)(2,160)(3,159)...(79,83)(80,82)(81)$

$\sigma \tau^{161} = \sigma = (1,161)(2,160)(3,159)...(79,83)(80,82)(81)$

$\sigma^2 = e$


Раскраски задают одинаковые бусы тогда и только тогда, когда они лежат в одной орбите под действием $G$. Таким образом, задача состоит в том, чтобы вычислить количество орбит действия $G$ на $X$. Для этого будем применять лемму Бернсайда.

Чтобы ею воспользоваться, необходимо предварительно найти - сколько элементов множества $X$ оставляют на месте каждый элемент группы $G$. 


Для подсчёта таких элементов множества $X$ нужно узнать количество независимых циклов. Для этого воспользуемся леммой 3.

161 раскладывается на простые множители 7 и 23. 
Следовательно НОД(161, i) = 7, если i кратно 7; = 23, если i кратно 23; = 1, если i не кратно ни 7, ни 23.


$|X^{e}| = 6^{161}$, длина = 161

$|X^{\tau^{k}}| = 6^{23}$, где $k \in \{ 23, 46, 69, 92, 115, 138 \}$, длина = 6

$|X^{\tau^{l}}| = 6^{7}$, где $l \in \{ 7, 14, 21, 28, 35, 42, 49, 56, 63, 70, 77, 84, 91, 98, 105, 112, 119, 126, 133, 140, 147, 154 \}$, длина = 22

$|X^{\tau^{k}}| = 6^{1}$, где $k \in \{ 1, 2, 3, ..., 161 \} \setminus$ (k $\cup$ l)$, длина = 132

$|X^{\sigma}| = |X^{\sigma \tau}| = |X^{\sigma \tau^{2}}| = ... = |X^{\sigma \tau^{160}}| = 6^{81}$

$|G| = 322$

Количество циклов при использовании $\sigma$ = 81. При использовании композиции $\sigma$ и $\tau^i$ количество циклов тоже = 81 из за симметрии структуры перестановки. Сколько бы поворотов ни было сделано, при применении осевой симметрии один элемент всегда будет на своём месте, а другие будут образовавыть циклы длины 2.

Использовав лемму Бернсайда, количество различных бус (число орбит) будет равно:
