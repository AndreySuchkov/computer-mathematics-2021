## Основные теоретические положения

**def1**

Пусть следующие элементы это:

$\mathbb{G}$ - группа

$e$ - нейтральный элемент

$\mathbb{X}$ - множество.

$\mathbb{G}$ действует на $\mathbb{X}$, если $\mathbb{G} \times \mathbb{X} \rightarrow \mathbb{X}$ такая, что $\forall x \in \mathbb{X}$, $\forall g, h \in \mathbb{G}$ выполняется:

$1)\ ex = x $

$2)\ g(hx)=(gh)x$


**def2**

Орбитой $x \in \mathbb{X}$ под действием группы $\mathbb{G}$ называется: $Gx = \{gx| g \in \mathbb{G}\} = Orb(x)$. Длина орбиты - количество элементов в ней.

**def3**

Неподвижные точки элемента $g \in \mathbb{G}$ - это такие $x \in \mathbb{X}$, что $gx = x$. Обозначение множества неподвижных точек: $\mathbb{X}^g$.

**def4**

Пусть $I$ - произвольное множество, $\mathbb{C}$ - множество цветов. Раскраской множетсва $I$ называется функция $I\rightarrow\mathbb{G}$. $\mathbb{C}^I$ - множество всех раскрасок. $\# \mathbb{C}^I=(\#\mathbb{C})^{\#I}$

**def5**

Путь $\mathbb{G}$ - группа, действующая на $I$, тогда $gf(i) = f(g^{-1}(i))$, где $g \in \mathbb{G}$, $i \in I$, $f \in \mathbb{C}^I$.

**def6**

Количество раскрасок $\mathbb{C}^I$, которые сохраняют элемент $g \in \mathbb{G}$ равно $(\# \mathbb{C})^{\mathbb{C}(\alpha(g))}$

**Теорема**
Для элемента $g \in \mathbb G$ и функции $f \in \mathbb C ^I$ следующие условия эквивалентны:

1) $gf=f$

2) $f(i) = f(g^n i) ,\forall i \in I , n \in \mathbb Z$

3) Если $i,j \in I$ лежат в одном цикле в циклической записи перестановки $\varphi (x)$, то $f(i) = f(j)$.

**Лемма Бёрнсайда**

Количество орбит действия $\mathbb{G}$ по $\mathbb{X}$ равно: $N= \frac{1}{_{\#G}} \cdot \sum_{g \in G}\# X^g$

**Лемма 2**

Предположим, что $\sigma$ является циклом длины $k$, а НОД($k \, n$) $= d$. Тогда перестановка $\sigma ^ n $ является произведением d циклов длины $k/d$.

### Решение задачи аналитическим способом
Имеем  2 набора значений: 
1. набор бусинок $\{x_1, x_2,...,x_{55}\}$
2. набор цветов $\{a,b,c,d\}$
Каждая бусинка, $x_i$, окрашена в один из цветов набора цветов. Общее количество бус, которое может получиться - $4^{55}$. В таком наборе учитываются и повторяющиеся бусы.

На множестве бус определены две операции - поворот по ч.с.(бусинки смещаются на 1 позицию вперёд) и осевая симметрия (отражение бусинок относительно некторой оси).  Пусть $\tau$ - поворот на $\frac{360}{55}$ градусов по часовой стрелке. Пусть $\sigma$ - осевая симметрия кольца бус.

Чтобы воспользоваться леммой Бернсайда необходимо рассчитать сколько элементов множества $\mathbb{X}$ оставляют на месте каждый элемент группы $\mathbb{G}$: $\#\mathbb{X}^{e} = 4^{55}$

Разложим $55$ на простые множители: $55 = 11*5$. $\tau^i$ является произведением из $d$ независимых циклов, где $d$=НОД($55, i$). Тогда для всех $i$ не кратных $11$ или $5$ $d$ будет иметь значение $1$.

$\#\mathbb{X}^{\tau^i} = 4^{2}$ при зачении i не кратному 11 и 5.

$\#\mathbb{X}^{\tau^{11}} = \#\mathbb{X}^{\tau^{22}} = \#\mathbb{X}^{\tau^{33}} = ... = \#\mathbb{X}^{\tau^{55}} =  4^{11}$ при зачении i кратному 11.

$\#\mathbb{X}^{\sigma\tau} = \#\mathbb{X}^{\sigma\tau^2} = ... =  \#\mathbb{X}^{\sigma\tau^{54}} = 4^{28}$ 

Количество орбит действия $\mathbb{G}$ на $\mathbb{X}$ :

$\#\mathbb{G} = 110$
