## Основные теоритические положения

### Действие группы на множество 
Пусть $G$ - группа с нейтральным элементом $e$,  
$X$ - множество  
$G$ - действует на $X$, если задана операция $G\times X \rightarrow G$  
Обозначение: $gx$

    Cвойства:
1. $ex=x$
2. $g(hx)=(gh)x$




### Орбита элемента
Орбитой $x\in X$ под действием $G$ называется множество $Gx=\{gx|g\in G \}$  
Количество элементов в орбите называется длиной орбиты.

### Неподвижная точка элемента
Неподвижной точкой элемента $g\in G$ называется элемент $x \in X$ для которого выполняется условие: $gx = x$  
$X^g$ - множество неподвижных точек элемнта $g$

### Теорема Бернсайда
Количество орбит действия группы $G$ на $X$ равно:  
$N=\frac{1}{\# G}\Sigma_{g\in G}{\#} X^g$

### Множество всех раскрасок
Пусть $I$ - произвольное множество, $C$ - множество цветов, тогда раскраской множесетва $I$ нызвается функция из $I$ в $C$  
$C^I$ - множество всех раскрасок  
$\# C^I = (\# C)^{\# I}$  


Пусть $G$ - группа, действующая на $I$. Тогда формула $gf(i)=f(g^{-1}i)$ задает действие $G$ на $C^I$  
$g\in G,\ i\in I,\ f\in C^I$

$g\in G,\ f\in C^I$

1) $gf=f$  
2) $f(i)=f(g^ni),\ \ \forall i \in I,\ \ \forall n \in \mathbb{Z}$  
3) Если $i,j\in I$ лежат в одном цикле в циклической записи $\phi (g)$, то $f(i)=f(j)$

Количество раскрасок из $C^I$, которые сохраняют элемент $g\in G$ равно $(\# C)^{C(\phi (g))}$

### 1. Аналитическое решение задачи о раскрасках
Сначала разделим правильный шестиугольник на одинаковые треугольники и пронумеруем каждый.



Каждый треугольник - это элемент нашего множество, назовем это множество $X$. На множесте $X$ заданы 2 действия: $\phi$ - поворот шестиугольника на 60 градусов, $\tau$ - поворот по оси. Запишем все действия над множеством в циклической форме:

$e= (1)(2)(3)(4)(5)(6)$  
$\phi = (1\ 2\ 3\ 4\ 5\ 6)$  
$\phi^2 = (1\ 3\ 5)(2\ 4\ 6)$  
$\phi^3 = (1\ 4)(2\ 5)(3\ 6)$  
$\phi^4 = (1\ 5\ 3)(2\ 6\ 4)$  
$\phi^5 = (1\ 6\ 5\ 4\ 3\ 2)$  
$\tau = (1\ 6)(2\ 5)(3\ 4)$  
$\tau\phi = (1\ 5)(2\ 4)(3)(6)$  
$\tau\phi^2=(1\ 4)(2\ 3)(5\ 6)$  
$\tau\phi^3=(1\ 3)(2)(4\ 6)(5)$  
$\tau\phi^4=(1\ 2)(3\ 6)(4\ 5)$  
$\tau\phi^5=(1)(2\ 6)(3\ 5)(4)$  
Получили:  
$\# G=12$  
$\#X^e = (\#C)^6$  
$\#X^{\phi}= (\#C)$  
$\#X^{\phi^2}= (\#C)^2$  
$\#X^{\phi^3}= (\#C)^3$

### 2. Графическое отображение раскраски

```python
from sage.plot.graphics import Graphics
from sage.plot.polygon import polygon
import math as math
import random

def createTriangle(x,y,r, reverse=False):
    angles = []
    if reverse == False:
        angles.append((x,y))
        angles.append((x+r/2,y-math.sin(math.pi/3)*r))
        angles.append((x-r/2,y-math.sin(math.pi/3)*r))
    else:
        angles.append((x,y-math.sin(math.pi/3)*r))
        angles.append((x+r/2,y))
        angles.append((x-r/2,y))
    return angles

def createHexagon(number):
    hexagon = polygon(createTriangle(0,0,5),axes = false)
    x0 = 0
    x = 0
    y = 0
    r = 5
    colors = ['red','blue', 'green']
    numOfTriangles = [11,13,15,17,19]
    reverse = False
    for num in numOfTriangles:
        for i in range(num):
            hexagon = hexagon + polygon(createTriangle(x,y,r, reverse),axes = false, edgecolor='black',color=colors[number%3])
            number //= 3
            x = x + r/2
            reverse = (reverse + 1)%2
        x = x0
        x = x - r/2
        x0 = x
        y = y - math.sqrt(3)*r/2
        reverse = False
    numOfTriangles.reverse()
    reverse = True
    x = x + r/2
    x0 = x
    for num in numOfTriangles:
        for i in range(num):
            hexagon = hexagon + polygon(createTriangle(x,y,r, reverse),axes = false, edgecolor='black',color=colors[number%3])
            number //= 3
            x = x + r/2
            reverse = (reverse + 1)%2
        x = x0
        x = x + r/2
        x0 = x
        y = y - math.sqrt(3)*r/2
        reverse = True
    
    return hexagon
```
