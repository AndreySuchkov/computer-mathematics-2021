## Основные теоретические положения

Пусть $G$ - группа с нейтральным элементом $e$, $X$ - произволльное множество.
$G$ <b>действует</b> на $X$, если задана операция $G\times X \longrightarrow X$ , при этом образ пары $(g, x)$ обозначается $gx$. Операция обладает следующими свойствами:

$\forall x\in X, g, h\in G$:
1. $ex=x$
2. $g(hx)=(gh)x$

<b>Орбитой</b> элемента $x\in X$ под действием группы $G$ называется множество
$Gx=\{gx|g\in G\}$

<b>Множеством неподвижных точек</b> элемента $g\in G$ - называется множество
$X^g = \{x|x\in X, gx=x\}$

<b>Лемма 1 (Бёрнсайда):</b>
Количество обрит действия группы $G$ на множество $X$ вычисляется по следующей формуле:

$N=\large\frac{1}{(\#G)}*\sum\limits_{g\in G}(\#X^g)$

Пусть $I$ - произвольное множество, $C$ - множество цветов, тогда <b>раскраска</b> множества $I$ - это функция из $I$ в $C$:

$I\longrightarrow C: C^I$

$C^I$ - множество всех раскрасок

$\#C^I=(\#C)^{\large\#I}$

Пусть $G$ - группа, действующая на множество $I$, тогда $gf(i)=f(g^{-1}i)$ задаёт действие $G$ на множество $C^I$, где\
$g\in G, i\in I, f\in C^I$

<b>Лемма 2:</b>

Для элемента $g\in G$ и функции $f\in G^I$:
1. $gf=f\ \forall n\in \mathbb{Z}$
2. $f(i)=f(g^ni)\ \forall i\in I, n\in \mathbb{Z}$
3. Если $i,j\in I$ лежат в одном цикле циклической записи перестановки $\alpha(g)$, то $f(i)=f(j)$

Количество раскрасок $C^I$, <b>сохраняющих элемент</b> $g\in G$, равно $(\#C)^{\large c(\alpha(g))}$, где $c(\alpha(g))$ - количество циклов в записи перестановок $\alpha(g)$.

## Аналитическое решение задачи

Группа преобразований $G$ содержит следующие преобразования шестиугольника:

1. $e = \tau^0 = \tau^6 = \sigma^2$ = - тождественное преобразование
2. $\tau$ - поворот на $60^\circ$
3. $\tau^2$ - поворот на $120^\circ$
4. $\tau^3$ - поворот на $180^\circ$
5. $\tau^4$ - поворот на $240^\circ$
6. $\tau^5$ - поворот на $300^\circ$
7. $\sigma = \sigma\tau^6 = \sigma e$ - отражение по вертикальной оси симметрии (проходящей через середины оснований)
8. $\sigma\tau$ - отражение с поворотом на $60^\circ$
9. $\sigma\tau^2$ - отражение с поворотом на $120^\circ$
10. $\sigma\tau^3$ - отражение с поворотом на $180^\circ$
11. $\sigma\tau^4$ - отражение с поворотом на $240^\circ$
12. $\sigma\tau^5$ - отражение с поворотом на $300^\circ$

$G = \{e, \tau, \tau^2, \tau^3, \tau^4, \tau^5, \sigma, \sigma\tau, \sigma\tau^2, \sigma\tau^3, \sigma\tau^4, \sigma\tau^5\} ,\ \#G = 12$

Для отражения шестиугольника достаточно выбрать одну ось симметрии, поскольку отражения по всем остальным вкупе с поворотами и без них выражаются через отражения и отражения с поворотами по выбранной оси, что делает их эквивалентными.

<b>Замечание.</b> Линии, проведённые через вершины шестиугольника параллельно его основаниям, делят его на $6$ "больших" равных правильных треугольников. В одном таком треугольнике содержится $16$ "маленьких" равных правильных треугольников (то есть всего "маленьких" треугольников $96$), что даёт основания утверждать, что для всех них поворот соответствует переходу в следующий по направлению поворота "большой" треугольник, а отражение - переходу в диаметрально противоположный "большой" треугольник (с небольшими оговорками в тех случаях, когда при отражении или комбинации отражения и поворотов "большие" треугольники переходят сами в себя). Таким образом независимые циклы, справедливые для "больших" треугольников, справедливы также и для "маленьких", содержащихся в них, а это значит, что количество независимых циклов преобразования в большинстве случаев будет кратно 16.

По <i>лемме 2</i> количество раскрасок множества $I$ в цвета множества $C$, сохраняющих элемент $g\in G$, равно $(\#C)^{c(\alpha(g))}$, где $c(\alpha(g))$ - количество независимых циклов (определяющее количество неподвижных элементов элемента $g\in G$). Найдём количество неподвижных элементов для каждого $g\in G$.

1. $\normalsize\#X^e = 6^{96}$, так как для для всех "маленьких" треугольников длины циклов будут равны $1$, а это значит, что количество циклов равно количеству этих треугольников.
2. $\normalsize\#X^\tau = 6^{16}$, поскольку образуются циклы длины $6$ для каждых $6$ "маленьких" треугольников.
3. $\normalsize\#X^{\tau^2} = 6^{32}$, все циклы имеют длину $3$, после $3$ поворотов элементы возвращаются на исходную позицию.
4. $\normalsize\#X^{\tau^3} = 6^{48}$, рассуждения аналогичны пункту <b>3</b>, все циклы имеют длину $2$.
5. $\normalsize\#X^{\tau^4} = 6^{32}$, исходя из того, что осуществляется поворот на $120^\circ$, только в обратную сторону, т.е. все циклы имеют длину $3$
6. $\normalsize\#X^{\tau^5} = 6^{16}$, потому как совершается $6$ поворотов на $60^\circ$ в обратную сторону, и для каждых $6$ "маленьких" треугольников существует общий цикл, всего циклов $16$.
7. $\normalsize\#X^\sigma = 6^{52}$, потому что $8$ элементов мозаики, расположенных на оси симметрии, переходят сами в себя, а остальные $88$ элементов образуют пары (циклы длины $2$), значит всего циклов $8 + 88 / 2 = 8 + 44 = 52$ 
8. $\normalsize\#X^{\sigma\tau} = 6^{48}$, что следует из того, что для трёх пар "больших" треугольников элементы каждой из них переходят друг в друга, но поскольку выполняется поворот, так или иначе "маленькие" треугольники не переходят сами в себя, и для каждых двух "маленьких" треугольников существует один цикл, значит, всего циклов $48$.
9. $\normalsize\#X^{\sigma\tau^2} = 6^{52}$, этот случай представляет особый интерес. Элементы двух пар "больших" треугольников переходят друг в друга, поэтому для всех их элементов длины циклов равны $2$. Для этих $4$ "больших" треугольников количество циклов будет равно $64 / 2 = 32$. Помимо них есть ещё два треугольника, которые при преобразовании переходят в отражённую относительно высоты, проведённой из центра шестиугольника к его стороне, версию себя (можно считать это преобразование отражением по оси симметрии, которая соответствует повёрнутой на $60^\circ$ исходной оси симметрии). Центральные элементы этих треугольников (лежащие на обозначенной высоте) в этом случае не меняются, а все остальные переходят друг в друга (внутри содержащего их "большого" треугольника), хотя в этом случае после отражения и применяются повороты. Поэтому для этих двух особых треугольников количество независимых циклов будет равно $2*(4 + 12/2) = 2*10 = 20$. Всего циклов $32+20 = 52$.
10. $\normalsize\#X^{\sigma\tau^3} = 6^{48}$, ситуация аналогична пункту 8.
11. $\normalsize\#X^{\sigma\tau^4} = 6^{52}$, ситуация аналогична пункту 9.
12. $\normalsize\#X^{\sigma\tau^5} = 6^{48}$, ситуация аналогична пункту 8.

Заметим, что отражение в совокупности с поворотами также является отражением, но относительно других осей, являющихся повёрнутыми исходными осями, поэтому для отражений с чётным числом ($\{0,2,4\}$) поворотов, которые являются отражениями относительно осей, проходящих через середины сторон шестиугольника, число циклов равно $52$ (не меняются элементы, лежащие на оси симметрии), а для отражений с нечётным числом поворотов ($\{1,3,5\}$), являющихся отражениями относительно осей, проходящих через противоположные вершины, число циклов равно $48$.

Теперь, когда известно количество неподвижных элементов для всех $g\in G$, есть возможность найти общее количество раскрасок (орбит действия G), воспользовавшись <i>леммой 1 (Бёрнсайда)</i>:

$N = \large\frac{1}{(\#G)}*\sum\limits_{g\in G}(\#X^g)$

$N = \Large\frac{6^{96}+2*6^{16}+2*6^{32}+4*6^{48}+3*6^{52}}{12}$

## Отображение раскраски по заданному номеру

Поскольку необходимо выбрать цвет для каждого из $96$ треугольников, нужно при построении каждого из них выбирать для строящегося какой-то один цвет из шести. Можно на каждом шаге выбирать цвет, вычисляя остаток от деления заданного числа на количество цветов ($6$), а затем выполняя целочисленное деление на это число. Тогда для того, чтобы выбрать цвет для $i$-го по порядку отрисовки треугольника, необходимо разделить заданное число на $6^i$ (что соответствует делению на $6$ $i$ раз) и взять остаток от деления результата на $6$. В этом случае входной параметр можно задать в виде суммы степеней $6$, умноженных на коэффициенты, соответствующие индексам цветов из набора имеющихся. Эта сумма также будет иметь представление в схеме Горнера, где остатки от деления - это индексы выбираемого на каждом шаге цвета. При каждом последующем делении на $6$ будем получать очередной остаток, который и определит выбираемый цвет. Поскольку самый первый остаток не умножается на $6$, будем считать его умноженным на $6^0 = 1$, и степени будут принимать значения от $0$ до $95$.

Пример раскраски, в котором $6$, $35$ и $72$ треугольники покрашены в цвета, отличные от остальных (треугольники считаются сверху вниз по рядам, слева направо по столбцам):

$6^{5}(6^{29}(6^{37}*random(1, 5) + random(1,5)) + random(1,5))$

```python
from sage.plot.polygon import polygon as poly

randLayer1 = [randint(0,5) for i in range(42)]
randLayer2 = [randint(0,5) for i in range(30)]
randLayer3 = [randint(0,5) for i in range(18)]
randLayer4 = [randint(0,5) for i in range(6)]
randLayers = [randLayer1, randLayer2, randLayer3, randLayer4]

exampleLayers = [[0]*42, [0]*30, [0]*18, [0]*6]
exampleLayers[0][1] = randint(1,5)
exampleLayers[0][11] = randint(1,5)
exampleLayers[2][11] = randint(1,5)

def formPaint(layers):
    paint = []
    paint += layers[0][6:15]
    paint += layers[0][4:6] + layers[1][4:11] + layers[0][15:17]
    paint += layers[0][2:4] + layers[1][2:4] + layers[2][2:7] + layers[1][11:13] + layers[0][17:19]
    
    paint += layers[0][0:2] + layers[1][0:2] + layers[2][0:2] + layers[3][0:3]\
                + layers[2][7:9] + layers[1][13:15] + layers[0][19:21]
    
    paint += layers[0][:-3:-1] + layers[1][:-3:-1] + layers[2][:-3:-1] + layers[3][:-4:-1]\
                + layers[2][-8:-10:-1] + layers[1][-14:-16:-1] + layers[0][-20:-22:-1]
    
    paint += layers[0][-3:-5:-1] + layers[1][-3:-5:-1] + layers[2][-3:-8:-1]\
                + layers[1][-12:-14:-1] + layers[0][-18:-20:-1]
    
    paint += layers[0][-5:-7:-1] + layers[1][-5:-12:-1] + layers[0][-16:-18:-1]
    paint += layers[0][-7:-16:-1]
    
    for i in range(len(paint)):
        paint[i] *= 6**i
    
    return paint

def rotate(layers, num):
    layers[0] = layers[0][(7*num)%len(layers[0]):] + layers[0][:(7*num)%len(layers[0])]
    layers[1] = layers[1][(5*num)%len(layers[1]):] + layers[1][:(5*num)%len(layers[1])]
    layers[2] = layers[2][(3*num)%len(layers[2]):] + layers[2][:(3*num)%len(layers[2])]
    layers[3] = layers[3][(1*num)%len(layers[3]):] + layers[3][:(1*num)%len(layers[3])]
    return layers

def flip(layers):
    for i in range(len(layers)):
        length = len(layers[i])//2
        for j in range(length//2):
            layers[i][j], layers[i][length-j-1] = layers[i][length-j-1], layers[i][j]
            layers[i][-j-1], layers[i][length+j] = layers[i][length+j], layers[i][-j-1]
    return layers

def createTriangle(x, y, col, flipped=False):
    if flipped:
        return poly([(x, y), (x+0.5, y+0.5*sqrt(3)), (x-0.5, y+0.5*sqrt(3))] , color=col, edgecolor=(0.1, 0.1, 0.1))
    return poly([(x, y), (x+1, y), (x+0.5, y+0.5*sqrt(3))] , color=col, edgecolor=(0.1, 0.1, 0.1))

def drawHex(param):
    colors = ((1, 0.32, 0.41), (1, 0.1, 0.67), (1, 0.63, 0.07), (0.36, 1, 0.68), (0.69, 1, 0.24), (0.31, 0.38, 1))
    hexagon = poly([(0, 0)], axes = False)
    leftX, y = -2.5, 1.5*sqrt(3)
    Qrows = tuple(9+2*i for i in range(4))
    for q in Qrows:
        flip = False
        x = leftX
        for i in range(q):
            hexagon += createTriangle(x, y, colors[param%len(colors)], flip)
            flip = not flip
            if flip:
                x += 1
            param //= len(colors)
        y -= 0.5*sqrt(3)
        leftX -= 0.5
    leftX += 1
    for q in reversed(Qrows):
        flip = True
        x = leftX
        for i in range(q):
            hexagon += createTriangle(x, y, colors[param%len(colors)], flip)
            flip = not flip
            if flip:
                x += 1
            param //= len(colors)
        y -= 0.5*sqrt(3)
        leftX += 0.5
    return hexagon
```

```python
print("Пример раскраски, данный выше:")
example = formPaint(exampleLayers)
exampleHexagon = drawHex(sum(example))
exampleHexagon.show()

print("Та же раскраска, только отражённая:")
exampleLayers = flip(exampleLayers)
example = formPaint(exampleLayers)
exampleHexagon = drawHex(sum(example))
exampleHexagon.show()

num = 2
print("Поворот отражённой раскраски на {} градусов (влево):".format(60*num))
exampleLayers = rotate(exampleLayers, num)
example = formPaint(exampleLayers)
exampleHexagon = drawHex(sum(example))
exampleHexagon.show()

print("Случайная раскраска:")
randomPaint = formPaint(randLayers)
randomHexagon = drawHex(sum(randomPaint))
randomHexagon.show()
```
