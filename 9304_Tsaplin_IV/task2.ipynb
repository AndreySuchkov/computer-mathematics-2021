```python
plot = Graphics()
colors = ["red", "cyan","yellow", "blue", "magenta", "green"]
numberOfCuts = 6

def make_hex(number):
    global colors
    global numberOfCuts
    hex = []
    for i in range(numberOfCuts^2 * 6):
        hex.append(colors[number % len(colors)])
        number //= 6
    return hex
    
    
def draw_tile(x, y, orientation, colorPar):
    global plot
    if (orientation == 0):
        plot += polygon2d([[x,y], [x - 0.5,y - sqrt(3)/2], [x + 0.5,y - sqrt(3)/2]], color = colorPar, thickness = 1, edgecolor = "black")
    elif (orientation == 1):
        plot += polygon2d([[x,y], [x - 0.5,y + sqrt(3)/2], [x + 0.5,y + sqrt(3)/2]], color = colorPar, thickness = 1, edgecolor = "black")

def draw_hex(hex):
    n = sqrt(len(hex)/6)*2+1
    flag = 0
    x = 0
    y = 0
    cur = 0
    global plot
    
    for i in range(sqrt(len(hex)/6)):
        for j in range(n):
            draw_tile(x, y, flag, hex[cur])
            cur += 1
        
            if(flag == 0):
                x += 0.5
                y -= sqrt(3)/2
            elif(flag == 1):
                x += 0.5
                y += sqrt(3)/2
            
            flag = (flag + 1) % 2
        
        x = x - n/2 - 0.5
        flag = 0
        n += 2
    
    flag = 1
    x += 0.5
    y -= sqrt(3)/2
    
    while(n > sqrt(len(hex)/6)*2+2):
        n -= 2
        for j in range(n):
            draw_tile(x, y, flag, hex[cur])
            cur += 1
        
            if(flag == 0):
                x += 0.5
                y -= sqrt(3)/2
            elif(flag == 1):
                x += 0.5
                y += sqrt(3)/2
            
            flag = (flag + 1) % 2
        x = x - n/2 + 0.5
        y -= sqrt(3)
        flag = 1
    

draw_hex(make_hex(randint(0, len(colors)**(numberOfCuts^2*6))))
plot.show(axes = false)
```
