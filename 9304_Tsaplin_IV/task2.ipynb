### Основные теоретические положения
**Определение 1.** Пусть $G$ - группа с нейтральным элементом *e*, а $X$ - множество. Будем говорить, что $G$ действует на $X$, если задана операция $G \times X \rightarrow X$ (образ пары $(g, x)$ обозначается обычно просто $gx$), обладающая для любого $x \in X$ и $g,h \in G$ следующими свойствами:

    1) ex = x
    2) g(hx) = (gh)x
    
**Определение 2.** Пусть X - множество. Множество всех биективных функций $X \rightarrow X$ с операцией композиции называется *симметрической группой* на множестве $X$ и обозначается через $S_X$

**Лемма-Определение 3.** Орбитой элемента $x \in X$ под действием $G$ называется множество $Gx = \{ gx \vert g \in G \}$. Количество элементов в данной орбите называется длиной орбиты.<br>
Любые две орбиты либо не пересекаются, либо совпадают. Таким образом, множество $X$ разбивается в дизъюнктивное объединение орбит.

**Определение 4.** Неподвижными точками элемента $g \in G$ называются те $x \in X$, для которых $gx = x$. Множество неподвижных точек элемента g обозначается через $X^g$.

**Определение 5.** Множество элементов группы $G$, оставляющих на месте данный элемент $x \in X$ называется *стабилизатором* элемента $x$ и обозначается через $St(x)$. Другими словами, $St(x) = \{g \in G \vert gx = x\}$. Очевидно, что стабилизатор является подгруппой в $G$.

**Замечание 1.** Количество пар $(g, x) \in G \times X$, для которых $gx = x$ можно вычислить двумя способами, которые указаны в разных частях следующего неравенства:<br>
$$
\begin{aligned}
 \sum_{x \in X}|{St(x)|} = \sum_{g \in G}|X^g|
\end{aligned}
$$

**Лемма 1.** Длина орбиты элемента $x$ равна индексу стабилизатора этого элемента. Поэтому, если группа $G$ конечна, то $|Gx| \cdot |St(x)| = |G|$.

**Лемма Бернсайда** Количество орбит действия группы G на множестве X равно<br>
$$
\begin{aligned}
 \cfrac{1}{|G|}\sum_{g \in G}|X^g|
\end{aligned}
$$

### Реализация функции отображения раскраски по заданному номеру.

1)make_hex() - функция получает номер и по нему составляет список цветов, в которые раскрашены треугольники в шестиугольнике<br>
2)draw_tile() - рисует треугольник заданного цвета. Если параметр orientation равен 1, рисует перевёрнутый треугольник<br>
3)draw_hex() - по списку цветов рисует раскрашенный шестиугольник<br>

```python
plot = Graphics()
colors = ["red", "cyan","yellow", "blue", "magenta", "green"]
numberOfCuts = 6

def make_hex(number):
    global colors
    global numberOfCuts
    hex = []
    for i in range(numberOfCuts^2 * 6):
        hex.append(colors[number % len(colors)])
        number //= 6
    return hex
    
    
def draw_tile(x, y, orientation, colorPar):
    global plot
    if (orientation == 0):
        plot += polygon2d([[x,y], [x - 0.5,y - sqrt(3)/2], [x + 0.5,y - sqrt(3)/2]], color = colorPar, thickness = 1, edgecolor = "black")
    elif (orientation == 1):
        plot += polygon2d([[x,y], [x - 0.5,y + sqrt(3)/2], [x + 0.5,y + sqrt(3)/2]], color = colorPar, thickness = 1, edgecolor = "black")

def draw_hex(hex):
    n = sqrt(len(hex)/6)*2+1
    flag = 0
    x = 0
    y = 0
    cur = 0
    global plot
    
    for i in range(sqrt(len(hex)/6)):
        for j in range(n):
            draw_tile(x, y, flag, hex[cur])
            cur += 1
        
            if(flag == 0):
                x += 0.5
                y -= sqrt(3)/2
            elif(flag == 1):
                x += 0.5
                y += sqrt(3)/2
            
            flag = (flag + 1) % 2
        
        x = x - n/2 - 0.5
        flag = 0
        n += 2
    
    flag = 1
    x += 0.5
    y -= sqrt(3)/2
    
    while(n > sqrt(len(hex)/6)*2+2):
        n -= 2
        for j in range(n):
            draw_tile(x, y, flag, hex[cur])
            cur += 1
        
            if(flag == 0):
                x += 0.5
                y -= sqrt(3)/2
            elif(flag == 1):
                x += 0.5
                y += sqrt(3)/2
            
            flag = (flag + 1) % 2
        x = x - n/2 + 0.5
        y -= sqrt(3)
        flag = 1
    

draw_hex(make_hex(randint(0, len(colors)**(numberOfCuts^2*6))))
plot.show(axes = false)
```

Каждое новое деление стороны, добавляет новый слой треугольников, которые формируют вложенные шестиугольники. На рисунке 2 представлена схема шестиугольника, на которой вложенные шестиугольники выделены красным цветом.

Каждый новый вложенный шестиугольник добавляет в перестановку поворота цепочку, которая состоит из $2n - 1$ множителей, где $n$ - номер вложенного шестиугольника. Каждый множитель - перестановка, которая содержит столько независимых циклов, сколько содержит переставновка данного поворота, для шестиугольника, состоящего из 6 треугольников, то есть первого шестиугольника.


Для симметрий, которые не оставляют треугольников на месте, число независимых циклов будет равно половине числа треугольников.


Для симметрий, которые оставляют на месте треугольники, которые лежат на линии симметрии, число независимых циклов будет равно сумме числа треугольников, которые лежат на линии симметрии, и половине числа оставшихся треугольников.


Число треугольников, которые лежат на линии симметрии, равно $2n$, то есть удвоенному числу вложенных шестиугольников


Таким образом, для каждого преобразования можно рассчитать число независимых циклов:

$\tau: 1 \cdot\sum_{i = 1}^{n}(2i-1) = n^2$<br>


$\tau^2: 2 \cdot \sum_{i = 1}^{n}(2i-1) = 2 \cdot n^2$<br>


$\tau^3: 3 \cdot \sum_{i = 1}^{n}(2i-1) = 3 \cdot n^2$<br>


$\tau^4: 2 \cdot \sum_{i = 1}^{n}(2i-1) = 2 \cdot n^2$<br>


$\tau^5: 1 \cdot \sum_{i = 1}^{n}(2i-1) = n^2$<br>


$id: 6 \cdot \sum_{i = 1}^{n}(2i-1) = 6 \cdot n^2$ - общее число треугольников в шестиугольнике<br>


$S_1: 2\cdot n +((6 \cdot n^2- 2\cdot n)/2) = 3 \cdot n^2 + n$<br>


$S_2: (6\cdot n^2)/2 = 3 \cdot n^2$<br>


$S_3: 2\cdot n +((6 \cdot n^2- 2\cdot n)/2) = 3 \cdot n^2 + n$<br>


$S_4: (6\cdot n^2)/2 = 3 \cdot n^2$<br>


$S_5: 2\cdot n +((6 \cdot n^2- 2\cdot n)/2) = 3 \cdot n^2 + n$<br>


$S_6: (6\cdot n^2)/2 = 3 \cdot n^2$<br>


Таким образом, общая формула для числа шестиугольников $N$ будет иметь следующий вид:

$N = \cfrac{1}{12}(|C|^{6 \cdot n^2} + 3 \cdot |C|^{3 \cdot n^2 + n} + 4 \cdot |C|^{3 \cdot n^2} + 2 \cdot |C|^{2 \cdot n^2} + 2 \cdot |C|^{n^2})$, 

где $n$ - число частей, на которые разбита сторона шестиугольника, С - множество цветов
