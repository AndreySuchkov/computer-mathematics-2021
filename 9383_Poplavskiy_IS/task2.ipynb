#### Лемма Бёрнсайда
Количество орбит действия группы $\mathbb G$ на множестве $\mathbb X$ равно $$N = \frac{1}{|G|}\sum_{g \in G} |X^{g}|$$

Сколькими способами можно составить бусы из 62 бусинок 3 различных цветов (с точностью до поворотов и осевых симметрий)?

**Решение**

По условию имеем множество цветов {$a, b, c$} и множество бусин {$x_1, x_2, ..., x_{61}$} $\forall i\  x_i \in \{a,b,c\}$. Всего существует $3^{62}$ раскрасок, но не все являются уникальными.

Существуют две такие операции, что бусы до их применения, а также после остаются одиннаковыми:
* При повороте
* При симметричном отображение 

Следовательно на множестве $X$ всех раскрасок действует группа поворотов и осевых симметрий. Операцией в этой группе является композиция, она является ассоциативной, нейтральный элемент - тождественное отображение, обратным к осевой симметрии является она сама, обратным к повроту - поворот в обратную сторону. Обозначим данную группу как $G$.

Через $\tau$ обозначим поворот, переводящий каждое место в следующее, последний элемент переходит на первый.

Через $\sigma$ обозначим осевую симметрию.

Найдем сколько циклов в каждой перестановке, которые получаются под действием группы $G$. Для этого воспользуемся Леммой 1, при помощи которой операцией $\tau$ найдем количество циклов перестановок.

Например для перестановки {1, 2, 3, ..., 62} после применения операции поворота получится перестановка {62, 1, 2, ..., 61}. Соответствующий цикл для данной перестановке (1, 2, 3, ..., 62) c длиной равной k = 62.

Т.е. для следующих вычислений количества циклов в перестановках $\tau^{k}$ будем использовать НОД(n, k) = d.

Число $62$ раскладывается на простые множители как $62 = 2*31$. Каждая перестановка $\tau^k$ является произведением $d$ независимых циклов, где $d= НОД(62, k)$. Следовательно для всех $k$ не кратных 2 или 31 $d$ будет равняться 1. 

Расмотрим операцию $\sigma$, как она работает:
* Последний элемент переходит в первый 
* Предпоследний элемент переход во второй
* ...
* Так как у нас четное число, то посередине не останется элемента и 31-й, 32-й элемент поменяются местами

Операции $\sigma$ соответствует перестановка (1, 62)(2, 61)...(31, 32)
Количество циклов = 31

Вывод по операциям: раскраски совмещенные каким-нибудь преобразованием из $G$ задают одинаковые наборы (бусы), т.е. когда они лежат в одной орбите под действием $G$.

Используя лемму Бёрнсайда и лемму 1, посчитаем ответ:

$|X^{e}| = 3^{62}$

$|X^{\tau^{i}}| = 3$, где $i \in \{1..61\} / \{ i | (i\%2 = 0) || (i\%31 = 0)\}$

$|X^{\tau^{k}}| = 3^{2}$, $\forall k \in \{2,4,6,8,...,60\}$

$|X^{\tau^{l}}| = 3^{31}$

$|X^{\sigma}| = |X^{\sigma \tau}| = |X^{\sigma \tau^{2}}| = ... = 3^{31}$

$|G| = 124$

$$\frac{1}{124}(3^{62} + 30*3^1 +3^{2}*30 + 3^{31} + 3^{31}*62)$$
