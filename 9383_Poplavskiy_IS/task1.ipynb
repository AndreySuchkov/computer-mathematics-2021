### Основные теоретические положения
#### Бинарный метод
Бинарное (двоичное) возведение в степень — это приём, позволяющий возводить любое число в n-ую степень за O(log n) умножений (вместо n умножений при обычном подходе). В данном методе степень представляется в бинарном виде.
* Если бит равен 0, то основание возводится в квадрат.
* Если бит равен 1, то число возводится в квадарт, а затем умножается само на себя.
* Количество операций рассчитывается по формуле: $\lambda (n)+ \nu(n) - 1$, где $\lambda (n)$ - уменьшенная на 1 длина двоичной записи числа n ($\left\lfloor\log_2 n \right\rfloor$)
* $\nu(n)$ - вес хэминга (число отличных от 0 цифр в записи числа).

#### Аддитивная цепочка
Последовательность натуральных чисел, начинающаяся с единицы, в которой каждый последующий элемент является суммой каких-то двух предшествующих элементов (в том числе, возможно использование одного и того же предшествующего элемента — удвоение). Формально, в аддитивной последовательности $\alpha_i$ выполнены условия:
* $\alpha_0$ = 1
* для любого $i > 0, \alpha_i = \alpha_j + \alpha_k$, где $j, k < i$

Одной из практически интересных разновидностей аддитивной цепочки является цепочка, заканчивающаяся числом $n$ в которой каждый последующий элемент является удвоением предыдущего или суммой предыдущего и первого элементов:
* для любого $i > 0, \alpha_i = \alpha_{i-1} + \alpha_{i-1}$ или $\alpha_i = \alpha_{i-1} + \alpha_0$

#### Метод Брауэра
позволяет по заданной линейной цепочке $1, а_1,..., а_m = n$ построить аддитивную цепочку для $2n — 1$ следующим образом.
Выписывается последовательность: <br> 
$$ 1 = 2^1 - 1, 2^{a_1} - 1, ..., 2^{a_m} - 1 = 2^{n} - 1 $$
затем в промежутки между числами вставляются последовательности удвоений. Так, между соседними числами $2^{\alpha_i}-1$, где $\alpha_{i+1} = \alpha_i + \alpha_j$, помещается последовательность 
$$2(2^{\alpha_i}-1),2^2(2^{a_i}-1),...,2^{\alpha_{i+1}-\alpha_i}(2^{\alpha_i}-1)=2^{\alpha_{i+1}}-2^{\alpha_j}.$$
Поскольку
$$2^{\alpha_{i+1}}-1=(2^{\alpha_i+1}-2^{a_j})+(2^{\alpha_j}-1)$$
итоговая последовательность является линейной аддитивной цепочкой длины $n+m-1$ состоящей из $n — 1$ удвоений и $m$ сложений

*Метод Брауэра* можно использовать и с произвольной цепочкой для $n$. При этом, если $\alpha_i = \alpha_j + \alpha_k$, то вычисления $2^{\alpha_i}-1$ последовательность удвоений необходимо вести от $2^{\alpha_j}-1$ или от $2^{\alpha_k}-1$. Число сложений в построенной цепочке также будет совпадать
с длиной исходной цепочки для $n$, , но число удвоений может быть больше, чем $n-1$. Цепочки, для которых методом Брауэра строится цепочка ровно с $n-1$ удвоениями, называются цепочками Ханзена. В цепочке, построенной методом Брауэра с исходной цепочкой Ханзена,
все нелинейные шаги являются удвоениями.

### Сравнение бинарного метода и метода множителей

##### Реализация бинарного метода

```python
def binpow(num, degree):
    bin_pow = bin(degree)[3:]
    opperations = 0
    result = num
    for bit in bin_pow:
        if bit == '0':
            result *= result
            opperations += 1
        else:
            result *= result 
            result *= num
            opperations += 2
            
    return result, opperations
```

##### Реализация метода множителей

```python
def multiplier(num, degree):
    factors = factor(degree)
    if factors[0][0] == degree:
        result, op = binpow(num, degree-1)
        return result * num, op + 1
    
    a = binpow(num, factors[0][0])
    b = binpow(a[0], degree // factors[0][0])

    return b[0], a[1] + b[1]
```

### 2. Алгоритм Брауэра

```python
def brauer(n, k, chain):
    if n < 2**k:
        for i in range(1, 2**k):
            chain.append(i)    
    else:
        q = n // 2**k
        chain = brauer(q, k, chain)
        for i in range(1, k+1):
            q *= 2
            chain.append(q)    
        chain.append(n)
        
    return chain
```

### 3. Алгоритм дробления вектора индексов

```python
def build_chain(r_i, ro_i):
    q = len(r_i)
    chain = [1]
    for index in r_i:
        chain.append(chain[-1] + chain[index - 1])
    for index in ro_i:
        chain.append(chain[-1] + chain[index - 1])
    return chain
    
def inc_vector(ind_vec, start_index=1):
    max_index = start_index + len(ind_vec) - 1
    for i in range(1, len(ind_vec) + 1):
        ind_vec[-i] += 1
        if ind_vec[-i] > max_index:
            if i == len(ind_vec):
                ind_vec.append(1)
                break
            ind_vec[-i] = 1
            max_index -= 1
        else:
            break
            
#разложение ветора индексов        
def split_vector(n):
    m = int(log(n, 2))
    up_border = m + bin(n).count('1') - 1
    if m == log(n, 2):
        m -= 1
    while m <= up_border:
        q = int(m/2)
        if q == 0:
            q = 1
        r_i = [1 for i in range(q)]
        while len(r_i) == q:
            ro_i = [1 for i in range(m - q)]
            chain = build_chain(r_i, ro_i)
            if chain[-1] == n:
                return chain
            a_max = chain[q] * 2 ** (m - q)
            a_min = chain[q] + m - q
            if n < a_min or n > a_max:
                inc_vector(r_i)
                continue
            while len(ro_i) == m - q:
                chain = build_chain(r_i, ro_i)
                if chain[-1] == n:
                    return chain
                inc_vector(ro_i, q + 1)
            inc_vector(r_i)
        m += 1
```

```python
import time 
for i in [502, 535, 560]:
    start = time.time()
    vect = split_vector(i)
    end = time.time() - start
    print("Цепочка для n = {} построенна за {} c, цепочка {}, длина методом Брауэра = {}\n".format(i, end, vect, len(vect), brauer(i, 2, [])))
```
