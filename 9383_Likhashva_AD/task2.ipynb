## Основные теоретические положения

#### Определение 1

Пусть $G$ - группа, $e$ - нейтральный элемент группы $G$, $X$ - множество. $G$ действует на $X$, если задана операция $G x X -> G$.
Тогда $\forall x \in X$, $\forall g,h \in G$: 
1) $ex=x$

2) $g(hx)=(gh)x$ 

#### Определение 2

Пусть $X$ - множество. Множество всех биективных функций с $X \to X$ с операцией композиции называется симметрической группой на множестве $X$ и обозначается как $S_{X}$

#### Определение 3

Орбитой $x \in  X $ под действием группы $G$ называется множество $Gx = \{gx | g \in G\}$. Количество элементов в данной орбите называется длиной орбиты.

#### Определение 4

Неподвижными точками элемента $g \in G$ называют те $x \in  X$ для которых выполнено : $gx = x$.

Обозначение множества неподвижных точек элемента $g$: $ X^g$

#### Определение 5

Пусть $I$ - произвольное множество, а $C$ - множество цветов. Раскраской множества $I$ называется функция из $I$ в $C$. $C^{I}$ - множество всех раскрасок.

#### Определение 6

Пусть $G$ группа, действующая на множестве $I$. Тогда $gf(i) = f(g^{-1}i)$ задает действие $G$ на $C^{I}$, где $g \in G$, $i \in I$, $f \in C^{I}$.

#### Лемма Бёрнсайда

Количество орбит действия группы $G$ на множество $X$ вычисляется по формуле: $N= \frac{1}{_{\#G}} \cdot \sum_{g \in G}\# X^g $

#### Теорема

Для элемента $g \in G$ и функции $f \in C^{I}$ следующие условия эквивалентны:

1)$gf = f$

2)$f(i) = f(g^{n}i)$, $\forall i \in I $ и $n \in \mathbb{Z}$

3)Если $i,j \in I$ лежат в одном цикле в циклической записи перестановки $\varphi(g)$, то $f(i) = f(j)$ (в терминологии задачи о бусах, это означает, что все места из одного цикла покрашены в одинаковый цвет)

#### Определение 7

Количество раскрасок из $C^I$, которые сохраняют элемент $g \in G$ равно $|C|^{c(\alpha(g))}$, где $c(\alpha)$ - количество независимых циклов в перестановке $\alpha$.

#### Один поворот

```python
PrintBeads(Rotate(beads1), 6, 1)
PrintBeads(Rotate(Rotate(beads2)), 12, 2)
PrintBeads(Rotate(Rotate(Rotate(beads3))), 18, 3)
PrintBeads(Rotate(Rotate(Rotate(Rotate(beads4)))), 24, 4)
PrintBeads(Rotate(Rotate(Rotate(Rotate(Rotate(beads5))))), 30, 5)
```

```python
PrintBeads(Symmetry(beads1), 6, 1)
PrintBeads(Symmetry(beads2), 12, 2)
PrintBeads(Symmetry(beads3), 18, 3)
PrintBeads(Symmetry(beads4), 24, 4)
PrintBeads(Symmetry(beads5), 30, 5)
```
