# Основные теоретические положения

#### Опр. 1
Пусть $\mathbb G$ - группа с нейтральным элементом $e$, $\mathbb{X}$ - множество.

$\mathbb G$ действует на $\mathbb{X}$, если задана операция $\mathbb G  \times \mathbb{X} \to \mathbb G$

Свойства операции:

$\forall x \in \mathbb{X} , \forall g,h \in \mathbb G$$$1)\ ex = x $$$$2)\ g(hx)=(gh)x$$

#### Опр. 2
Орбитой $x \in   \mathbb{X} $ под действием группы $\mathbb G$ называется множество $\mathbb Gx = \{gx | g \in \mathbb G\}$.

#### Опр. 3
Длина орбиты - количество элементов в данной орбите.

#### Опр. 4
Неподвижными точками элемента $g \in \mathbb G$ называют те $x \in \mathbb X$ для которых выполнено : $gx = x$.

Обозначение множества неподвижных точек элемента $g$: $\mathbb  X^g$

#### Опр. 5
Пусть $I $ - произвольное множество, а $\mathbb C$ - множество цветов. Раскраской множество $ I$ называется функция из $ I$ в $\mathbb C$ .

Множество всех раскрасок: $\mathbb C ^{ I}$.

Количество раскрасок $= (\# \mathbb C)^{\#  I}$

#### Опр. 6
Пусть $\mathbb G $ - группа, действующая на множестве $I$ . Тогда формула

$$gf(i)=f(g^{-1}i),$$
где $g \in \mathbb G$, $i \in  I$, а $f \in \mathbb C ^{ I},$ задает действие $g$ на $\mathbb C^{ I}$.

#### Теорема
Для элемента $g \in \mathbb G$ и функции $f \in \mathbb C ^I$ следующие условия эквивалентны:

1) $gf=f$

2) $f(i) = f(g^n i) ,\forall i \in I , n \in \mathbb Z$

3) Если $i,j \in I$ лежат в одном цикле в циклической записи перестановки $\varphi (x)$, то $f(i) = f(j)$. (В терминологии задачи о бусах это означает, что все места из одного цикла покрашены в одинаковый цвет).

#### Опр. 7
Количество раскрасок из $\mathbb C^I$, которые сохранят данный элемент $g \in G $ равно $(\# \mathbb C)^{c(\varphi (g))}$. $c(\varphi)$ - количество независимых циклов в перестановке $\varphi$.

#### Лемма
Предположим, что $\sigma$ является циклом длины $k$, а НОД($k \, n$) $= d$. Тогда перестановка $\sigma ^ n $ является произведением d циклов длины $k/d$.

#### Лемма Бёрнсайда
Количество орбит действия группы $\mathbb G$ на множестве $\mathbb X$ равно$$N = \frac{1}{\# \mathbb G} * \sum_{ g \in \mathbb G} \# \mathbb X^g$$

### Задание
Сколькими способами можно составить бусы из 209 бусинок 4 различных цветов (с точностью до поворотов и осевых симметрий)?


### Решение

Пусть есть упорядоченный набор $\{ x_{1}, x_{2}, x_{3}, ... , x_{209} \}$, где $x_{i} \in \{ a, b, c, d \}$. Т.е. 209 бусинок, каждая из которых может быть раскрашена в один из 4 различных цветов. Существует всего $4^{209}$ таких раскрасок, но так как необходимо учесть повороты и симметрии, то бусы будут повторяться.

###### Поворот
Например, на первом месте зеленый шар, а на всех остальных белые, применяем поворот. Получим на втором месте зеленый шар, а на всех остальных белые. Такие бусы будем считать одинаковыми.

###### Симметрия
Если есть бусы $\{1, 2, 3, 4, 5, 6, 7, 8\}$, то симметричное отображения будет менять местами бусинки $1-8, 2-7, 3-6, 4-5 $, тогда получится буса $\{8, 7, 6, 5, 4, 3, 2, 1\}$. Такие бусы также будем считать одинаковыми.

Итак, на множестве всех раскрасок $X$ действует группа поворотов и осевых симметрий. Операция в группе - композиция, она является ассоциативной, нейтральный элемент - тождественное отображение, обратный элемент к осевой симметрии - сама симметрия (отобразили симметрично $\{1, 2, 3, 4\}$, получили $\{4, 3, 2, 1\}$ затем еще раз и получили исходную перестановку $\{1, 2, 3, 4\}$), обратный элемент к повороту - поворот в другую сторону. Обозначим эту группу как $G$.

Группа $G$ действует на множестве мест $\{1, 2, 3, ..., 209\}$, задано отображение из $G$ в симметрическую группу $S_{209}$. Пусть $\tau$ - поворот, переводящий каждое место в следующее(двигаем вправо), а последний элемент передвигаем на первый. А $\sigma$ - осевая симметрию относительно оси проходящую между 1 и 209, и 105.

Для перестановки $\{ 1, 2, 3, ... ,209 \}$ после применения поворота(операции $\tau$) получается перестановка $\{ 209, 1, 2, 3, ..., 208 \}$. Цикл соответствующий этой перестановке будет $(1,2,3,4,...,209)$ и его длина равна $k = 209$. Теперь для нахождения количества циклов в перестановках $\tau^{n}$ будет применяться формула $НОД(k,n) = d$.

 $209$  = $11 * 19$ , то есть 209 раскладывается на постые множители 11 и 19. И $НОД$ отличный от единицы будет получаться при $n \in \{ 11, 22, 33, 44, 55, 66, 77, 88, 99, 110, 121, 132, 143, 154, 165, 176, 187, 198, 19, 38, 57, 76, 95, 114, 133, 152, 171, 190\}$ Тогда $d$ будет принимать значения: $D = (d_{11},d_{22},...,d_{198},d_{19},...,d_{190})$

Рассмотрим перестановку после операции $\sigma$. Мы переводим последний элемент в первый, предпоследний во второй, ... а элемент по середине оставляем на месте. Т.е. этой операции будет соответствовать перестановка $(1, 209)(2, 208)(3, 207)...(105)$, количество циклов равно 105. После применении к этой перестановке операций вида $\tau^{k}$ не изменят количество циклов. После такого симметричного отображения для каждого последующего поворота у нас всегда найдется элемент, который на следующем повороте "встанет" на свое место. Т.е. в самом первом повороте встанет на место число 1, потому что он сейчас в самом конце, в последующем двойка и.т.д.

Раскраски, совмещаемые друг с другом каким-нибудь преобразованием из группы $G$ задают одинаковые бусы. Тогда можно сказать, что раскраски задают одинаковые бусы тогда и только тогда, когда они лежат в одной орбите под действием $G$. Таким образом, наша задача состоит в том, чтобы вычислить количество орбит действия $G$ на $X$.

Для того, чтобы воспользоваться леммой Бёрнсайда, необходимо выяснить, сколько элементов множества $X$ оставляет на месте каждый элемент группы $G$. Для подсчёта таких элементов множества $X$ нужно знать количество независимых циклов:

$|X^{e}| = 4^{209}$

$|X^{\tau^{i}}| = 4^{1}$, где $i \in \{1..208\} / \{ i | (i\%11 = 0) || (i\%19 = 0)\}$

$|X^{\tau^{11}}| = |X^{\tau^{22}}| = |X^{\tau^{33}}| = ... = |X^{\tau^{198}}| = 4^{11}$

$|X^{\tau^{19}}| = |X^{\tau^{38}}| = ... = |X^{\tau^{190}}| = 4^{19}$

$|X^{\sigma}| = |X^{\sigma \tau}| = |X^{\sigma \tau^{2}}| = ... = |X^{\sigma \tau^{208}}| = 4^{105}$

$|G| = 418$

##### По лемме Бёрнсайда количество различных бус (число орбит) равно:

$\frac{1}{418}(4^{209} + 180*4^{1} + 18*4^{11} + 10*4^{19} + 209*4^{105})$
