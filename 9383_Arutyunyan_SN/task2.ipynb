## Основные теоретические положения

#### Определение 1

Пусть $G$ - группа с нейтральным элементом $e$, а $X$ - произвольное множество. Тогда $G$ **действует** на X, если задана операция $\phi: G \times X \rightarrow X$. При этом пара $G \times X$ обозначается как $gx, g \in G, x \in X$.

Свойства операции $\phi$ $\forall x \in X, \forall g \in G$:

1. $ex = x$

2. $g(hx) = (gh)x$

#### Определение 2

**Орбитой** $x \in X$ под действием $G$ называется множество $Gx = \{gx | g \in G\}$.

Тогда $|Gx|$ - **длина орбиты**.

#### Определение 3

**Неподвижной точкой** элемента $g \in G$ называют те $x \in X$, для которых $gx = x$.

Далее в качестве $X^g$ будем понимать множество неподвижных точек для элемента $g \in G$.

### Лемма Бернсайда

Количество орбит N действия $G$ на $X$ задается следующей формулой: $N = \frac{1}{|G|}\sum_{g \in G} |X^g|$

#### Определение 4

Пусть $I$ - произвольное множество, а $C$ - множество цветов. Тогда **раскраской** множества $I$ называется функция $f: I \rightarrow C$.

#### Обозначение

Множество всех раскрасок обозначается как $C^I$. Тогда $|C^I| = |C|^{|I|}$.

#### Определение 5

Пусть $G$ - группа, действующая на множество $I$. Тогда формула $gf(i) = f(g^{-1}i)$, где $g \in G, i \in I, f \in C^I$, задает действие $G$ на множество $C^I$.

#### Утверждение

Пусть $g \in G, f \in C^I$. Тогда для этих $g$ и $f$ следующие условия эквивалентны:

1. $gf = f$

2. $f(i) = f(g^ni) \forall i \in I, \forall n \in Z$.

3. Если $i, j \in I$ лежат в одном цикле в циклической записи перестановки $\alpha(g)$, то $f(i) = f(g)$.

#### Определение 6

Количество раскрасок из $C^I$, которые сохраняют элемент $g \in G$ равно $|C|^{c(\alpha(g))}$, где $c(\alpha)$ - количество независимых циклов в перестановке $\alpha$.

## Графическое изображение раскрасок

```python
from sage.plot.polygon import polygon, polygon2d
import random

SIDE = 3
HEIGHT = SIDE * 3**(1/2)/2
COLORS = ["red", "green", "blue", "yellow"]

def make_triangle(x, y, color, reverse):
    return polygon([(x, y), (x + SIDE, y), (x + SIDE/2, y + HEIGHT)], color=color, edgecolor="black") if not reverse \
      else polygon([(x, y), (x - SIDE/2, y + HEIGHT), (x + SIDE/2, y + HEIGHT)], color=color, edgecolor="black")

def make_hexagon_half(startx, starty, ydirection, start_reverse, amounts, identifier):
    reverse = start_reverse
    hexagon_half = polygon([0, 0], axes=False)
    
    for amount in amounts:
        x, y = startx, starty
        for i in range(amount):
            hexagon_half += make_triangle(x, y, COLORS[identifier % len(COLORS)], reverse)
            if not reverse:
                x += SIDE

            reverse = not reverse
            identifier //= 3
            
        startx += SIDE / 2
        starty += ydirection * HEIGHT
        reverse = start_reverse
        
    return (hexagon_half, identifier)

def make_hexagon(identifier):
    hexagon = polygon([0, 0], axes=False)
    reverse = False
    triangles = [19, 17, 15, 13, 11]
    
    half, new_identifier = make_hexagon_half(-19, 0, 1, False, triangles, identifier)
    hexagon += half
    hexagon += make_hexagon_half(-19 + SIDE/2, -HEIGHT, -1, True, triangles, new_identifier)[0]

    return hexagon
```

```python
hexagon = make_hexagon(random.randint(0, 3**150))
hexagon.show()
```

## Решение задачи в общем виде

Если обобщать количество возможных цветов, то количество возможных раскрасок будет равно:
$N = \frac{2 \cdot A + 2 \cdot A^2 + 4 \cdot A^3 + 3 \cdot A^4 + A^6}{12}$, где $A$ - количество цветов.

Теперь попробуем обобщить количество частей, на которые делится сторона правильного шестиугольника.

Во-первых, количество частей, образованных его диагоналями, всегда равно 6. Во-вторых, можно доказать, что количество треугольников, на которые разбивается каждая из этих частей, равно $k^2$, где $k$ - количество частей, на которые делится сторона правильного шестиугольника.

Тогда количество раскрасок каждой части, образованной диагоналями, будет равно $A^{k^2}$. Тогда общее количество возможных раскрасок равно $N = \frac{2 \cdot A^{k^2} + 2 \cdot {A^{k^2}}^2 + 4 \cdot {A^{k^2}}^3 + 3 \cdot {A^{k^2}}^4 + {A^{k^2}}^6}{12} = \frac{2 \cdot A^{k^2} + 2 \cdot A^{k^4} + 4 \cdot A^{k^6} + 3 \cdot A^{k^8} + A^{k^{12}}}{12}$.
