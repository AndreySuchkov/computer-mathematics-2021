#### Лемма Бёрнсайда

Количество орбит действия группы $\mathbb G$ на множестве $\mathbb X$ равно: $N = \frac{1}{\# \mathbb G} * \sum_{ g \in \mathbb G} \# \mathbb X^g$

Множество всех раскрасок: $\mathbb C ^{ I}$.

Количество раскрасок $= (\# \mathbb C)^{\#  I}$

#### Определение 6

$gf(i)=f(g^{-1}i),$     где $g \in \mathbb G$, $i \in  I$, а $f \in \mathbb C ^{ I},$  задает действие $g$ на $\mathbb C^{ I}$. 

#### Теорема 

Для элемента $g \in \mathbb G$ и функции $f \in \mathbb C ^I$ следующие условия эквивалентны:

1) $gf=f$

2) $f(i) = f(g^n i) ,\forall i \in I , n \in \mathbb Z$

3) Если $i,j \in I$  лежат в одном цикле в циклической записи перестановки $\varphi (x)$, то  $f(i) = f(j)$. (В терминологии задачи о бусах это означает, что все места из одного цикла покрашены в одинаковый цвет).

#### Задание

Сколькими способами можно составить бусы из 93 бусинок 3 различных цветов (с точностью до поворотов и осевых симметрий)?

#### Решение
Дан упорядоченный набор $\{ x_{1}, x_{2}, x_{3}, ... , x_{93} \}$, где $x_{i} \in \{ a, b, c\}$. Т.е. 93 бусинки, каждая из которых может быть раскрашена в 3 различных цвета. Существует всего $3^{93}$ таких раскрасок, но не все из них соответствуют различным бусам.

На множестве $X$ всех раскрасок действует группа поворотов и осевых симметрий. Операцией в этой группе является композиция. Операция является ассоциативной, нейтральный элемент - тождественное отображение, обратным к осевой симметрии является она сама, обратным к повроту - поворот в обратную сторону. Обозначим группу буквой $G$.

Задано отображение из $G$ в симметрическую группу $S_{93}$. Пусть $\tau$ - поворот, переводящий каждое место в следующее. $\sigma$ - осевая симметрия.

Раскраски, совмещаемые друг с другом каким-нибудь преобразованием из группы $G$ задают одинаковые бусы. Т.е. раскраски задают одинаковые бусы, когда они лежат в одной орбите под действием $G$. Таким образом задача состоит в том, чтобы вычислить количество орбит действия $G$ на $X$.

Выпишем некоторые перестановки из $S_{93}$ в которые переходят элементы группы $G$:

$e =  \begin{pmatrix}
  1 & 2 & 3 & ... & 93\\
  1 & 2 & 3 & ... & 93
 \end{pmatrix} = (1)(2)(3)...(93)$ 
 
$\tau = \begin{pmatrix}
  1 & 2 & 3 & ... & 93\\
  2 & 3 & 4 & ... & 1
 \end{pmatrix} = (1,2,3,...,93)$
 
 $\tau^2 = \begin{pmatrix}
  1 & 2 & 3 & ... & 92 & 93\\
  3 & 4 & 5 & ... & 1   & 2
 \end{pmatrix} = (1,3,5,...,91,93,2,4,...,92)$
 
 $\sigma = \begin{pmatrix}
  1 & 2 & 3 & ... & 93\\
  93 & 92 & 91 & ... & 1
 \end{pmatrix} = (1,93)(2,92)(3,91)...(45,49)(46,48)(47)$
 

$\sigma^2 = e$ 

Перестановка $\tau$ является циклом длины 93, исходя из Леммы 2 можно посчитать количество циклов в перестановке $\tau^i\ \forall i$. Число $93$ раскладывается на простые множители как $93 = 3*31$. Каждая перестановка $\tau^i$ является произведением $d$ независимых циклов, где $d:= НОД(93,i)$.

d = НОД(93, i) может принимать значения:

- 1, когда i не кратна ни 3, ни 31;
- 3, когда i кратна 3, но не кратна 31;
- 31, когда i кратна 31, но не кратна 3.

Теперь из определения 7 мы можем найти количество раскрасок, которые сохраняет данный элемент $g \in G$:

$\# X^{e} = 3^{93}$

$\# X^{\tau^{m}} \ = \ 3^{1}$, m = {1,2,4,5,7,...,91,92}, $\#m = 60$

$\# X^{\tau^{p}} \ = \ 3^{3}$, p = {3,6,9,...,90}, $\#p = 30 $

$\# X^{\tau^{q}} \ = \ 3^{31}$, q = {31,62}, $\#q = 2$

$\# X^{\sigma} \ = \ 3^{47}$

Перестановка $\sigma$ является произведением 47 независимых циклов. Рассмотрим из скольки циклов состоят перестановки вида $\sigma \tau^i$.

$\tau^i$ переводит элемент на месте $k$ на место $((k+i-1)\ mod\ 93 + 1)$. 

$\sigma $ переводит элемент $k$ в $(94-k)$. 

$\sigma \tau^i$ переводит $k$ в $(93 - (k+i-1)\ mod\ 93)$ 

$\sigma \tau^i$ переводит $(93-(k+i-1)\ mod \ 93)$ в $k$.

Т.е. перестановки $\sigma \tau^i$ либо переводит элемент сам в себя, либо переводит 2 элемента друг в друга. 

Чтобы понять сколько элементов переходят сами в себя можно составить уравнение $k= 93 - (k+i-1) \ mod \ 93$. 

$k = (46 -\frac{i}{2}) \ mod \ 93 + 1$.  

$(\frac{1}{2} \ mod \ 93 = 47)$.

Следовательно только один элемент перейдет сам в себя, а остальные 92 элементов будут образовывать циклы длины 2 с какими-нибудь другими элементами. Значит количество независимых циклов в перестановке $\sigma \tau^i$ = 47.

$\#G = 186$

Теперь мы можем воспользоваться леммой Бернсайда: количество различных бус равно $$\frac{1}{186}(3^{93} + 3*60 +3^{3}*30 + 3^{31}*2 + 3^{47}*93)$$
