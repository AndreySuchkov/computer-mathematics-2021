## Основные теоретические положения

__Действие группы на множество__

_Def:_ $\; \sqsupset G$ - группа с нейтральным элементом $e$.

$\qquad X$ - множество.

$G$ действует на $X$, если задана операция: $G\times X \rightarrow X$ (записывемая $gx$, где $x\in X$, $g \in G$), 
которая имеет следующие свойства:

<ol>
<li> $ex = x$, $\forall x \in X$. </li>
<li> $g(hx) = (gh)x$, $\forall x \in X$, $\forall g \in G,\,h \in G$. </li>
</ol>  

__Орбита элемента__

_Def:_ Орбита элемента $x \in X$ под действием группы $G$ - множество $Gx = \left\{gx \; \big| \; g \in G\right\}$.

Длина орбиты - количество элементов в данной орбите.

__Неподвижные точки__

_Def:_ Неподвижные точки элемента $g \in G$ - $X^g$ множество $x \in X$, таких что $gx = x$.


__Лемма Бёрнсайда__
 
 _Теорема:_$\;$Количество орбит действия группы $G$ на множество $X$ равно: 
 <center> $ N = \dfrac{1}{|G|}\displaystyle\sum\limits_{g \in G} |X^g| $ </center>

 _Доказательство_
 
 Каждый элемент $x \in X$ лежит в орбите $G_x$. Примем мощность каждой орбиты за единицу $|G_X| \sim 1$. 
 Тогда $\forall x_i \in G_x$ сопоставим число $\dfrac{1}{|G_x|}$. Тогда $\displaystyle\sum\limits_{x_i \in G_x} \dfrac{1}{|G_x|} = 1$. $\displaystyle\sum\limits_{x \in X} \dfrac{1}{|G_x|} = \displaystyle\sum\limits_{x_i \in G_x} \dfrac{1}{|G_x|} + \displaystyle\sum\limits_{x_i \in G_y} \dfrac{1}{|G_y|} + \dots = N$. Так как длина орбиты любого элемента равна индексу его стабилизатора, то $|G_x| = \dfrac{|G|}{|St(x)|}$. Следовательно, $\displaystyle\sum\limits_{x \in X} |St(x)|/|G| = \dfrac{1}{|G|}\displaystyle\sum\limits_{x \in X} |St(x)| = N$. По свойству $\displaystyle\sum\limits_{x \in X} |St(x)| = \displaystyle\sum\limits_{g \in G} |X^g|$. Следовательно лемма доказана.
 
 __Постановка задачи о раскрасках__
 
 $\sqsupset I$ - произвольное множество, $С$ - множество цветов.
 
 _Def:_ Раскраска множества $I$ - функция $I \rightarrow C$
 
 $C^I$ - множество всех раскрасок.

$|C^I| = |C|^{|I|}$

Пусть группа  G  действует на множество  I , тогда формула $gf(i)=f(g^{−1}(i)),g \in G,i \in I,f \in C^I$
задаёт действие $G$ на множество  $C^I$.

Для элемента $g \in G$ и функции $f \in C^I$ следующие условия эквивалентны:
<ol>
<li>$gf = f$</li>

<li>$f(i) = f(g^{n}i)$, $\forall i \in I $ и $n \in \mathbb{Z}$</li>

<li>Если $i,j \in I$ лежат в одном цикле в циклической записи перестановки $\varphi(g)$, то $f(i) = f(j)$</li>
</ol>

 _Def:_ Количество раскрасок из множетсва $C^I$, которые сохраняют элемент $g \in G$ равно $|C|^{c(\phi(g))}$, $c(\phi(g))$ - число независимых циклов перестановки $\phi(g)$

 _Доказательство_
 
Пусть $ \left< g \right> $ - циклическая подгруппа, порождённая элементом $ g $, обозначим $ \left< g \right> \! i $ орбитой элемента $ i $, $ Q $ - множество различных орбит. 

Так как $ gf = f $, то все элементы одной орбиты отображаются в одно и то же под действием функции $ f $. Поэтому $ f $ однозначно определяется функцией $ \tilde{f} : Q \to C $, где $ f \left( \left< g \right> \! i \right) = f \left( i \right) $, это значит, что множество $ \left( C^I \right)^{g} $ биективно отображается на множество $ C^Q $, следовательно, $ |\left( C^I \right)^{g}| = |C^Q| $. 

Количество всех функций из $ Q $ в $ C $ равно $ |C|^{|Q|} = |C|^{c \left( \varphi \left( g \right) \right)} $, что и требовалось доказать.

$G$ - группа преобразованй множества.

Для простоты обозначения преобразований пронумеруем грани куба.

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import numpy as np
from itertools import product, combinations

fig = plt.figure()
ax = fig.gca(projection='3d')
size = 10

M = np.array([[0,0,0],[size,0,0],[0,size,0],[size,size,0],[size,size,size],[0,0,size],[size,0,size],[0,size,size]])

c = 1
for s, e in combinations(M, 2):
    if np.sum(np.abs(s-e)) == size:
        ax.plot3D(*zip(s, e), color='b')
        c+=1
count = 0
z = 0

while z <= size:
        count += 1
        ax.text((size/2+z)%(size+1),0,z, count, fontsize = 20)
        count += 1
        ax.text(0,(size/2-z)%(size+1),z, count, fontsize = 20)
        count += 1
        ax.text((size/2-z)%(size+1),size,z, count, fontsize = 20)
        count += 1
        ax.text(size,(size/2+z)%(size+1),z, count, fontsize = 20)
        z += size/2


ris = plt.gcf()
ris.set_size_inches(20, 20, forward=True)

plt.axis('off')
plt.grid(b=None)
```

### Реализовать отображение раскраски куба по заданному номеру

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
from itertools import product, combinations

def cubeDone():
    n = int(input("Введите число раскраски:\n"))
    listOfColors = []
    n = abs(n)
    for i  in range(12):
        listOfColors.append(n%6)
        n = n//6
    size = 10
    M = np.array([[0,size,0],[size,0,0],[size,size,0],[0,0,0],[size,size,size],[size,0,size],[0,size,size],[0,0,size]])

    cube = []
    count = 0
    for s, e in combinations(M, 2):
        if np.sum(np.abs(s-e)) == size:
            cube.append([s,e,listOfColors[count]])
            count+=1
    return cube

def printColoring(cube):
    fig = plt.figure()
    ax = fig.gca(projection='3d')

    clr = ['#FF0000', '#00FF00', '#0000FF', '#FFFF00', '#FF00FF', '#00FFFF']

    count = 0
    for i in range(len(cube)):
        if np.sum(np.abs(cube[i][0] - cube[i][1])) == size:
            ax.plot3D(*zip(cube[i][0], cube[i][1]), color=clr[cube[i][2]])
            count+=1
    z = 0
    count = 0
    while z <= size:
        count += 1
        ax.text((size/2+z)%(size+1),0,z, count, fontsize = 20)
        count += 1
        ax.text(0,(size/2-z)%(size+1),z, count, fontsize = 20)
        count += 1
        ax.text((size/2-z)%(size+1),size,z, count, fontsize = 20)
        count += 1
        ax.text(size,(size/2+z)%(size+1),z, count, fontsize = 20)
        z += size/2

    ris = plt.gcf()
    ris.set_size_inches(20, 20, forward=True)

    plt.axis('off')
    plt.grid(b=None)
printColoring(cubeDone())
```
