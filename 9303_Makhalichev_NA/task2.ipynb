## Основные теоретические положения

Пусть $G$ группа с нейтральным элементом $e$, $X$ - множество. Группа $G$ действует на $X$, если задана операция $G \times X \rightarrow X$.

Свойства: ($\forall x \in X$, $\forall g, h \in G$)
1. $ex = x$
2. $g(hx) = (gh)x$

Орбитой элемента $x \in X$ под действием группы $G$ называется множество $Gx = \{gx | g \in G\}$. Количество элементов в данной орбите называется длиной орбиты.

Неподвижными точками элемента $g \in G$ называют те элементы $x \in X$, для которых выполняется $gx = x$.

$X^g$ - множество неподвижных точек элемента $g$.

Количество орбит действия группы $G$ на множество $X$ равно $N = \frac{1}{\#G} \sum_{g \in G}{\#X^g}$ (лемма Бёрнсайда).

Пусть $I$ - произвольное множество, $C$ - множество цветов, тогда раскраской множества $I$ называется функция из $I$ в $C$.

$C^I$ - множество всех раскрасок.

$\#C^I = {(\#C)}^{\#I}$.

Пусть $G$ - группа, действующая на множестве $I$, тогда формула $gf(i) = f(g^{-1}i)$ задаёт действие $G$ на множество $C^I$ ($g \in G$, $i \in I$, $f \in C^I$).

$g \in G$, $f \in C^I$. Для них эквивалентны следующие условия:
1. $gf = f$
2. $f(i) = f(g^ni)$ $\forall i \in I$, $\forall n \in \mathbb{Z}$
3. Если $i$, $j \in I$ лежат в одном цикле в циклической записи перестановки $\phi(g)$, то $f(i) = f(j)$.

Количество раскрасок из множества $C^I$, которые сохраняет элемент $g \in G$ равно ${(\#C)}^{c(\phi(g))}$ ($c$ - количество независимых циклов в записи перестановки $\phi(g)$).

Предположим, что $\sigma$ является циклом длины $k$, а НОД($k$, $n$)$ = d$. Тогда перестановка $\sigma^n$ является произведением $d$ циклов длины $\frac{k}{d}$.

### Реализация функций раскраски по заданному номеру

#### Функция поворота бус на одну бусину

```python
def Rotate(beads):
    return beads[1:] + beads[:1]
```

#### Функция симметричного отображения бус

```python
def Symmetry(beads):
    return beads[::-1]
```
