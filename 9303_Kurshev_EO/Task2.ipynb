## Основные теоретические сведения

### Действие группы на множестве

Говорят, что группа $G$ с нейтральным элементом $e$ действует на множестве $X$, если задана операция $G \times X \to G$ и она обладает следующими свойствами:<br>
$\forall x \in X, \forall g, h \in G$<br>
1. $ex = x$
2. $g(hx) = (gh)x$

### Орбита

Орбитой элемента $x \in X$ называется $Gx = \{gx|g \in G\}$

### Лемма Бёрнсайда

Количество орбит действия группы $G$ на множестве $X$ равно ${{1}\over|G|}\cdot {\sum\limits_{g \in G}}{|X^g|}$

### Раскраска

Пусть $l$ - произвольное множество, а $C$ - множество цветов. Тогда раскраской множества $l$ называется функция $l \to C$.<br>
$C^l$ - множество всех раскрасок

```python
import matplotlib.pyplot as plt
import numpy as np
from itertools import product, combinations

def cube_colors():
    n = int(input("Введите номер раскраски:\n"))
    Colors = []
    for i in range(12):
        Colors.append(n % 3)
        n = n // 3
    size = 1
    
    M = np.array([[0, 0, 0], [size, 0, 0], [0, size, 0], [size, size, 0], [size, size, size], [0, 0, size], [size, 0, size], [0, size, size]])
    cube = []
    count = 0
    for s, e in combinations(M, 2):
        if np.sum(np.abs(s - e)) == size:
            cube.append([s, e, Colors[count]])
            count += 1
    return cube

def print_cube(cube):
    fig = plt.figure()
    ax = fig.gca(projection = '3d')

    clr = ['#FF0000', '#00FF00', '#0000FF']

    for i in range(len(cube)):
        ax.plot3D(*zip(cube[i][0], cube[i][1]), color = clr[cube[i][2]])

    ris = plt.gcf()
    ris.set_size_inches(10, 10)

    plt.axis('off')
    plt.grid()
print_cube(cube_colors())
```
