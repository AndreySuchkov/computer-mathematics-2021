# Основные теоретические положения

## Бинарный метод
Бинарный метод возведения в степень - это алгоритм, который позволяет возводить любое число в натуральную степень n.  Происходит просмотр бит слева направо, не учитывая первый бит, так как он всегда равен единице: если бит равен единице, то текущее число возводится в квадрат и умножается на исходное чило, если бит равен нулю, то текущее число возводится в квадрат. Количество операций умножения в бинарном методе равно: $\lambda(n) + \nu(n) - 1$, где $\lambda(n) = \lfloor log_2 n \rfloor, \ \nu(n)$ - количество единиц в двоичной записи числа.

## Метод множителей возведения числа в в степень
Метод множителей возведения числа в в степень - это алгоритм, который позволяет возводить любое число в натуральную степень n. Если число простое, то это число возводится в степень $k-1$ и домножается на исходное. Если число составное, тогда степень числа представляется как произведение $k = i\cdot j$, где $i$ - наименьший простой множитель числа $k$. После число возводится в степень $i$, затем полученное число возводится в степень $j$.

## Аддитивная цепочка
Аддитивная цепочка для $n \in N$ - это последовательность $1=a_0, a_1, a_2, ... , a_m = n$, где $a_i = a_j + a_k, k \leq j &lt; i, \forall i = \overline{1...m}$. Длина минимальной аддитивной цепочки для числа $n$ равняется $l(n) = m$ Рекуррентная формула длины аддитивной цепочки: $l(n) = min\{l(n - 1) + 1, \ l_n\} + \sigma_n $

## Теорема Брауэра
Для числа $k &lt; log_2 (log_2 n)$ справедливо, что: $l(n) &lt; (1+k^{-1}) \cdot \lceil log_2 n \rceil + 2^{k-1} - k + 2$. Если $k=\lambda(\lambda(n))-2\lambda(\lambda(\lambda(n)))$, то следует что:

Следствие 1: $\lim \limits_{n \to \infty} \frac{l(n)}{\lambda(n)} = 1$

Следствие 2: Одна из наилучших верхних оценок длины аддитивной цепочки имеет вид: $\lambda(n)(1 + \frac{1}{\lambda(\lambda(n))} + \frac{o(\lambda(\lambda(\lambda(n))))}{\lambda(\lambda(n)) ^ 2})$


Основные теоретические положения
Бинарный метод
Основной идеей данного метода является разбиение искомого числа ${x^n}$ на произведение ${x}$ в некоторых степенях. Результат считается по следующим правилам:

Слева направо просматривается бинарная запись числа ${n}$ без первого бита. Пускай результ будет храниться в переменной ${t}$, которая изначально равняется ${x}$.

Если текущий бит - это ${1}$, то число возводится в квадрат, а полученное число затем еще раз умножается на ${x}$.

Если текущий бит - это ${0}$, то число возводится в квадрат. Количество операций умножения при использовании данного метода равно ${\lambda (n) + \nu (n) - 1}$, где ${\lambda (n) = \left \lfloor lb(n) \right \rfloor}$, а ${\nu(n)}$ - вес Хэмминга для бинарного представления числа ${n}$, то есть число ненулевых битов в двоичной записи числа

Метод множителей
Этот метод базируется на бинарном методе. Дано число ${x}$ и степень ${n}$, в которую это число предстоит возвести. Рассматриваем делители числа n, например, $n = a*b$, при этом $x^n =(x^a)^b$ . Тогда считаем $x^a$ бинарным методом, затем, делаем замену $y = x^a$, и считаем $y^b$.

Аддитивная цепочка
Аддитивная цепочка — это последовательность натуральных чисел ${a_0, a_1, a_2, ..., a_m = n}$, в которой первый элемент всегда равен ${1}$, а каждый последующий элемент является суммой каких-то двух предшествующих элементов (возможно, одинаковых - удвоение). То есть:

${a_0 = 1}$

${\forall i &gt; 0, a_i = a_j + a_k , j,k &lt; i}$ - шаг аддитивной цепочки.

Существуют следующие шаги:

1) ${i}$-шаг - удвоение, если ${j = k = i - 1}$

2) ${i}$-шаг - звездный, если ${j = i - 1}$, ${k = \overline{0 .. (i-1)}}$

3) ${i}$-шаг - малый, если ${\lambda(a_i) = \lambda(a_{i-1})}$

Алгоритм Яо для вычисления приближённых аддитивных цепочек
Цепочка Яо - аддитивная цепочка, состаящая из 2 аддитивных цепочек, первая цепочка равна $Y_k(n):1,2,4,\ ...\, 2^{\lambda (n)}$, вторая аддитивная цепочка составляется на основе первой. $d(z)$ - сумму всех разрядов, цифра которых равняется z.

Алгоритм Яо составляет для числа $n\in N$ цепочку Яо. Сначала алгоритму подается число $n\in N$ и $k\in N$. Далее алгоритм разбивает число на сумму: $\sum_{i=0}^{2^k-1}{2^{ik}a_i}$, где $a_i$ - цифра i-го разряда числа. Потом алгоритм считает $zd(z)$ для всех чисел от 1 до $2^k-1$. $zd(z)$ получается с помощью сложения чисел из $Y_k(n)$. Наконец сложением всех $d(z)$ получаем нужное нам число n. Все промежуточные значения записывются во вторую аддитивную цепочку цепочки Яо.

## Звездная цепочка
Звездная цепочка - это аддитивная цепочка, включающая только звездные шаги.

## Алгоритм дробления вектора индексов
Алгоритм дробления вектора индексов позволяет найти минимальную звездную цепочку для некоторого числа $n \in \mathbb{N}$.

Рассмотрим вектор индексов $\{r_i\}_{i=1}^q \cup {\{{\rho}_j \}}_{j=q+1}^m$, где ${\rho}_j= \{x: 1 \leq x \leq j \}$, ${\{r_i\}}_{i=1}^q$ - фиксированная часть, ${\{{\rho}_j\}}_{j=q+1}^m$ - изменяющаяся часть.

$a_max$ достигается при векторе индексов ${\{r_i\}}_{i=1}^{q} \cup \{q+1,q+2,\dots,m\}$.

$a_min$ достигается при векторе индексов ${\{r_i\}}_{i=1}^{q}\cup\{1,1,\dots,1\}$.

$a_{max} = a_{q+1} \cdot {2}^{m-q}$

$a_{min} = a_{q+1}+m-q$

Алгоритм:

1. Во внешнем цикле рассматриваем аддитивные цепочки длины $m$ от значения $\bar{l}(n)=\lceil log_2(n) \rceil$ до $\underline{l}(n)=\lambda(n)+\nu(n)-1$, на каждой итерации выбираем $q$ ($1 \leq q \leq m-1$);
2. Перебираем все возможные фиксированные части вектора индексов $\{r_i\}_{i=1}^q$ ($q!$ вариантов), для каждой строим соответствующую ей звездную цепочку, находим $a_{max}$ и ${a}_{min}$, после чего:
if $n \notin [a_{min},a_{max}]$, то переходим к следующему набору $\{r_i\}_{i=1}^q$;
if $n\in [a_{min},a_{max}]$, то перебираем все возможные изменяющиеся части вектора индексов ${\{{\rho}_j\}}_{j=q+1}^m$ и находим $a_m$:
if $a_m=n$, то цепочка найдена;
if все возможные изменяющиеся части вектора индексов ${\{{\rho}_j\}}_{j=q+1}^m$ исчерпаны, то переходим к следующему набору $\{r_i\}_{i=1}^q$;
if все наборы вектора индексов длины $m$ исчерпаны, то увеличиваем $m$ на 1.

# 1. Бинарный метод и метод множителей

```python
def binMethod(x, n):
    
    h = 0
    res = x
    b = bin(n)[3:]
    
    for i in b:
        h += 1
        if (i == '1'):
            res = res**2
            res = x*res
            h += 1
            
        else:
            res = res**2
            
    return (res, h)
```

```python
import math

def minM(n):
    
    for i in range(2, int(math.sqrt(n)) + 1):
        if (n%i == 0):
            return i
        
    return 1

def mulMethod(x, n):
    
    minm = minM(n)
    res1, h1 = binMethod(x, minm)
    res2, h2 = binMethod(res1, n // minm)
    
    return (res2, h1 + h2)
```

```python
x = 1
n = [33, 63, 205]

print("for x " + str(x))

for i in n:
    xbm, h_bm = binMethod(x,i) 
    x_mm, h_mm = mulMethod(x,i)
    
    print("Степень:"+str(i))
    print("MM " + str(h_bm))
    print("BM = " + str(h_mm))
```

Выбирать использование бинарного метода или метода множителей следует исходя из числа n. Если число n ближе к $2^k-1$, где $k \in N$, то стоит воспользоваться методом множителей. А если число n ближе к $2^k+1$, где $k \in N$, то лучше применить бинарный метод.

## 2. Алгоритм Брауэра

```python
from matplotlib import pyplot as plt
table_res = []

def Brauer(n, k, add):
    if(n < 2**k):
        for i in range(1, 2**k):
            add.append(i)
        return    
    else:
        q = n//(2**k)
        Brauer(q, k, add)
        for i in range(1, k+1):
            q *= 2 
            add.append(q)    
        add.append(n)
        
minChain = {}
n = [12, 32, 71, 190]
k = [2, 3, 4, 5]
for i in n:
    for j in k:
        add = []
        Brauer(i, j, add)
        print("n =", i, ", k =", j)
        print("цепочка: ", add)
        print("Длина: ", len(add))
        print()
        
        if minChain.get(i) is None or len(minChain[i][1]) >= len(add):
            minChain[i] = [j, add]
        
        row = [i, j, len(add)]
        table_res.append(row)
        
headers = ['n', 'k', 'Длина цепочки Брауэра']
cell_text = []
for i in table_res:
    cell_text.append([data for data in i])

table = plt.table(cell_text, rowLoc='right', colLabels=headers, loc='center')
table.set_fontsize(20)
table.scale(4, 2) 

ax = plt.gca()
ax.get_xaxis().set_visible(False)
ax.get_yaxis().set_visible(False)

plt.box(on=None)

plt.draw()
```

## 3. Алгоритм дробления вектора индексов

```python
import time
import math
from matplotlib import pyplot as plt

def binc(x):
    return bin(x)[3:].count('1')


def MChain(iV):
    chain = [1]
    for i in iV:
        chain.append(chain[-1] + chain[i - 1])
    return chain


def follow(arr, q = 0):
    if (arr == [1 for i in range(len(arr))]):
        return [0 for i in range(len(arr))]
    for i in range(len(arr) - 1, -1, -1):
        if (arr[i] == 1):
            arr[i] = q + i + 1
            continue
        arr[i] -= 1
        break
    return arr


def crash(n):
    for m in range(int(math.log(n, 2)), int(math.log(n, 2)) + binc(n) + 1):
        q = m // 2
        if (q == 0):
            q = 1
        r = [i for i in range(1, q + 1)]
        rh = [q + i + 1 for i in range(m - q)]
        while (r != [0 for i in range(q)]):
            chain = MChain(r + rh)
            cmin = chain[q] + (m - q)
            cmax = chain[q] * 2 ** (m - q)
            if (n < cmin):
                r = follow(r)
                continue
            elif (n > cmax):
                r = follow(r)
                continue
            if(chain[-1] == n):
                return chain
            while (rh != [0 for i in range(m - q)]):
                chain = MChain(r + rh)
                if (n == chain[-1]):
                    return chain
                rh = follow(rh, q)
            r = follow(r)
            rh = [q + i + 1 for i in range(m - q)]
    return []


res = []
arr = [12, 15, 501]
for n in arr:
    start = time.time()
    temp = crash(n)
    end = time.time()
    print('n =', n, ', l(n) =', len(temp))
    print('Потрачено времени:', end - start, ' секунд')
    print('Звездная цепочка:', temp)
    print()
    row = [n, len(temp), end - start, temp]
    res.append(row)
    
    
    
headers = ['n', 'l(n)', 'Время (сек)', 'Звездная цепочка'] 
cell_text = []
for i in res:
    cell_text.append([data for data in i])

table = plt.table(cell_text, rowLoc='right', colLabels=headers, loc='center')
table.set_fontsize(15)
table.auto_set_column_width(col=list(range(3)))
table.scale(5, 4) 

ax = plt.gca()
ax.get_xaxis().set_visible(False)
ax.get_yaxis().set_visible(False)

plt.box(on=None)

plt.draw()
```

Алгоритм дробления вектора индексов находит минимальную аддитивную цепочку для числа n, однако время работы программы очень большое.
