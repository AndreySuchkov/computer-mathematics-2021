Основные теоретические сведения
Действие группы на множестве

Говорят, что группа $G$ с нейтральным элементом $e$ действует на множестве $X$, если задана операция $G \times X \to X$ и она обладает следующими свойствами:

$\forall x \in X, \forall g, h \in G$

- $ex = x$

- $g(hx) = (gh)x$

## Орбита

Орбитой элемента $x \in X$ называется $Gx = \{gx|g \in G\}$
Лемма Бёрнсайда

Количество орбит действия группы $G$ на множестве $X$ равно ${{1}\over|G|}\cdot {\sum\limits_{g \in G}}{|X^g|}$
Раскраска

Пусть $l$ - произвольное множество, а $C$ - множество цветов. Тогда раскраской множества $l$ называется функция $l \to C$.
$C^l$ - множество всех раскрасок


## Лемма Бёрнсайда

Количество орбит действия группы $G$ на множестве $X$ равно ${{1}\over|G|}\cdot {\sum\limits_{g \in G}}{|X^g|}$

```python
from matplotlib import pyplot as plt

def draw_cube(n = 0):
    color = [
        "black",
        "red",
        "tan",
        "gold",
        "olive",
        "navy",
        "coral",
        "chocolate",
        ]
    number = []
    for i in range(8):
        number.append(color[n % 8])
        n = n // 6

    plt.hlines(y=5, xmin=0, xmax=5,color = "black")
    plt.hlines(y=0, xmin=0, xmax=5,color = "black")
    plt.vlines(x=5, ymin=0, ymax=5,color = "black")
    plt.vlines(x=0, ymin=0, ymax=5,color = "black")

    plt.hlines(y=7.5, xmin=2.5, xmax=7.5,color = "black")
    plt.hlines(y=2.5, xmin=2.5, xmax=7.5,color = "black")
    plt.vlines(x=7.5, ymin=2.5, ymax=7.5,color = "black")
    plt.vlines(x=2.5, ymin=2.5, ymax=7.5,color = "black")


    ax=plt.gca()

    plt.plot([0, 2.5], [0,2.5],color = "black")
    plt.plot([5, 7.5], [0, 2.5],color = "black")
    plt.plot([5, 7.5], [5, 7.5],color = "black")
    plt.plot([0, 2.5], [5, 7.5],color = "black")
    circle1 = plt.Circle((0, 0), 0.4, color=number[0])
    ax.text(0.0, 0.4, "1", color='black')
    circle2 = plt.Circle((2.5, 2.5), 0.4, color=number[1])
    ax.text(2.5, 2.9, "2", color='black')
    circle3 = plt.Circle((7.5, 2.5), 0.4, color=number[2])
    ax.text(7.5, 2.9, "3", color='black')
    circle4 = plt.Circle((5, 0), 0.4, color=number[3])
    ax.text(5.0, 0.4, "4", color='black')
    circle5 = plt.Circle((0, 5), 0.4, color=number[4])
    ax.text(0, 5.4, "5", color='black')
    circle6 = plt.Circle((2.5, 7.5), 0.4, color=number[5])
    ax.text(2.5, 7.9, "6", color='black')
    circle7 = plt.Circle((7.5, 7.5), 0.4, color=number[6])
    ax.text(7.5, 7.9, "7", color='black')
    circle8 = plt.Circle((5, 5), 0.4, color=number[7])
    ax.text(5.0, 5.4, "8", color='black')

    ax.add_patch(circle1)
    ax.add_patch(circle2)
    ax.add_patch(circle3)
    ax.add_patch(circle4)
    ax.add_patch(circle5)
    ax.add_patch(circle6)
    ax.add_patch(circle7)
    ax.add_patch(circle8)


    plt.grid()

    plt.axis('off')

    plt.show()


draw_cube(5555)
```
