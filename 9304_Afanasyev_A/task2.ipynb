## Основные теоретические положения

### Орбита

Пусть $G$ - группа с нейтральным элементом $e$, $X$ - множество.

$G$ действует на $X$, если задана операция: $G\times X \rightarrow X$, обозначаемая $gx$, где $x\in X$, $g \in G$

Свойства операции:
* $ex = x$, $\forall x \in X$
* $g(hx) = (gh)x$, $\forall x \in X$, $\forall g,\,h \in G$

Орбитой элемента $x \in X$ под действием $G$ называется множество $Orb(x) = \{ gx\,|\,g \in G\}$. Количество элементов в орбите называется длиной орбиты.

### Неподвижная точка

Неподвижными точками элемента $g \in G$ назвается множество $X^g = \{ gx = x\,|\,x \in X \}$.

### Теорема Бёрнсайда

Количество орбит действия $G$ на $X$ равно $N = \frac{1}{|G|}\sum_{g \in G} |X^g|$

### Раскраска

Пусть $I$ - множество, $C$ - множество цветов. Раскраской множества $I$ называется функция из $I$ в $C$. $C^I$ - множество всех раскрасок.

Количество раскрасок из $C^I$, которое сохраняет элемент $g \in G$ равно $|C|^{c(\phi(g))}$, где $c$ - количество независимых циклов в перестановке, включая циклы единичной длины.

Введем следующие обозначения:

* $\sigma_j^i$ - отражение вида j вида i раз

Для ясности рассуждений пронумеруем клетки квадрата следующим образом:

$$
\begin{array}{c|c|c|c|c}
    1 & 2 & 3 & 4 & 5 \\
    \hline
    6 & 7 & 8 & 9 & 10 \\
    \hline
    11 & 12 & 13 & 14 & 15 \\
    \hline
    16 & 17 & 18 & 19 & 20 \\
    \hline
    21 & 22 & 23 & 24 & 25 \\
\end{array}
$$

Тогда

$$
\begin{array}{l}
    \tau = (1,\,5,\,25,\,21)(2,\,10,\,24,\,16)\cdots(7,\,9,\,19,\,17)\cdots(13)) \\

    \tau^3 = (1,\,21,\,25,\,5)(2,\,16,\,24,\,10)\cdots(7,\,17,\,19,\,9)\cdots(13) \\

    \sigma_3 = (1)(2,\,6)(3,\,11)\cdots(25) \\
    \sigma_4 = (1,\,25)(2,\,20)(3,\,15)\cdots(21) \\
    \sigma_1^2 = \sigma_2^2 = \sigma_3^2 = \sigma_4^2 = e
\end{array}
$$

Отсюда получим, что $G = \{\tau,\,\tau^2,\,\tau^3,\,\sigma_1,\,\sigma_2,\,\sigma_3,\,\sigma_4,\,e\}$, $|G| = 8$.

В $\tau$ и в $\tau^3$ количество незивисимых циклов равно $\cfrac{25 - 1}{4} + 1 = 7$, так как длина каждого цикла, кроме одного, равна 4, и есть единственный элемент 13, который остался на своем месте.

В $\tau^2$ количество независимых циклов равно $\cfrac{25 - 1}{2} + 1 = 13$, так как длина всех циклов, кроме одного, равна 2, и снова только элемент 13 остался на своем месте.

В $\sigma_1$ количество независимых циклов равно $\cfrac{25 - 5}{2} + 5 = 15$, так как длина каждого цикла, кроме 5, равна 2, и 5 элементов на вертикальной оси остались на месте. Аналогично с $\sigma_2$.

В $\sigma_3$ количество независимых циклов равно $\cfrac{25 - 5}{2} + 5 = 15$, так как длина каждого цикла, кроме 5, равна 2, а 5 элементов на главной диагонали остались на месте. Аналогично с $\sigma_4$.

$$
|X^{e}| = 4^{25}, \; |X^{\tau}| = 4^{7}, \; |X^{\tau^2}| = 4^{13}, \; |X^{\tau^3}| = 4^{7}, \;
|X^{\sigma_1}| = 4^{15}, \; |X^{\sigma_2}| = 4^{15}, \; |X^{\sigma_3}| = 4^{15}. \; |X^{\sigma_4}| = 4^{15}
$$

Исходя из полученных значений, можно получить ответ с помощью теоремы Бёрнсайда:

$$
N = \frac{1}{|G|}\sum_{g \in G} |X^g| = \frac{1}{8}(4^{25} + 2 \cdot 4^{7} + 4^{13} + 4 \cdot 4^{15}) = 140738033618944
$$

### Реализация функций генерации, поворотов, отражений и отображения раскрашенного квадрата

#### Функции генерации:

```python
from random import randint

def generateNumber(colorsAmount, squareSize):
    number = 0
    for position in range(squareSize**2):
        number += colorsAmount**position * randint(0, colorsAmount - 1)
    return number


def createColoredSquare(colorsAmount, squareSize, number):
    square = []
    for rowIndex in range(squareSize):
        row = []
        for elementIndex in range(squareSize):
            row.append((rowIndex * squareSize + elementIndex + 1, number % colorsAmount))
            number //= colorsAmount
        square.append(row)
    return square


def createColoredSquareR(colorsAmount, squareSize):
    return [[(rowIndex * squareSize + elementIndex + 1, randint(0, colorsAmount - 1)) for elementIndex in range(squareSize)] for rowIndex in range(squareSize)]

    
```

#### Функции поворотов и отражений:

```python
def rotate90(square):
    for row in range(len(square) >> 1):
        for i in range(row, len(square) - row - 1):
            (square[i][len(square) - row - 1],
            square[len(square) - row - 1][len(square) - (i + 1)],
            square[len(square) - (i + 1)][row],
            square[row][i]) = (square[row][i], square[i][len(square) - row - 1],
            square[len(square) - row - 1][len(square) - (i + 1)],
            square[len(square) - (i + 1)][row])
    return square

  
def rotate180(square):
    for i in range((len(square)**2) >> 1):
        row = i // len(square)
        column = i % len(square)
        (square[row][column], 
        square[len(square) - row - 1][len(square) - column - 1]) = (square[len(square) - row - 1][len(square) - column - 1], 
        square[row][column])
    return square


def rotate270(square):
    for row in range(len(square) >> 1):
        for i in range(row, len(square) - row - 1):
            (square[row][i], square[i][len(square) - row - 1], 
            square[len(square) - row - 1][len(square) - (i + 1)],
            square[len(square) - (i + 1)][row]) = (square[i][len(square) - row - 1], 
            square[len(square) - row - 1][len(square) - (i + 1)],
            square[len(square) - (i + 1)][row], square[row][i])
    return square


def verticalReflect(square):
    for i in range(len(square)):
        for j in range(len(square) >> 1):
            (square[i][j],
            square[i][len(square) - (j + 1)]) = (square[i][len(square) - (j + 1)],
            square[i][j])
    return square


def horizontalReflect(square):
    for i in range(len(square) >> 1):
        for j in range(len(square)):
            (square[i][j],
            square[len(square) - (i + 1)][j]) = (square[len(square) - (i + 1)][j],
            square[i][j])
    return square


def diagonalReflect(square):
    for i in range(len(square)):
        for j in range(i, len(square)):
            (square[i][j],
            square[j][i]) = (square[j][i],
            square[i][j])
    return square


def subDiagonalReflect(square):
    for i in range(len(square)):
        for j in range(len(square) - i):
            (square[i][j],
            square[len(square) - (j + 1)][len(square) - (i + 1)]) = (square[len(square) - (j + 1)][len(square) - (i + 1)],
            square[i][j])
    return square
```

#### Функция отображения:

```python
from sage.plot.graphics import Graphics
from sage.plot.polygon import polygon
from sage.plot.text import Text

colors = ["grey", "orange", "red", "white", "blue", "green", "yellow", "pink"]

def printSquare(square):
    figure = Graphics()
    for i in range(len(square)):
        for j in range(len(square)):
            figure += polygon([[j, len(square) - i], [j + 1, len(square) - i], [j + 1, len(square) - (i + 1)], [j, len(square) - (i + 1)]], color=colors[square[i][j][1]], edgecolor="black")
            figure += text(square[i][j][0], (j + 0.5, len(square) - (i + 0.5)), color="black")
    figure.show(axes=False)
```

```python
squareSize = int(input("Enter a size of square: "))
colorsAmount = int(input("Enter a amount of colors: "))
# number = int(input("Enter a number: "), colorsAmount)
number = generateNumber(colorsAmount, squareSize)

print("Исходный квадрат:")
printSquare(createColoredSquare(colorsAmount, squareSize, number))

print("\nПоворот на 90 градусов:")
printSquare(rotate90(createColoredSquare(colorsAmount, squareSize, number)))

print("\nПоворот на 180 градусов:")
printSquare(rotate180(createColoredSquare(colorsAmount, squareSize, number)))

print("\nПоворот на 270 градусов:")
printSquare(rotate270(createColoredSquare(colorsAmount, squareSize, number)))

print("\nОтражение относительно вертикальной оси:")
printSquare(verticalReflect(createColoredSquare(colorsAmount, squareSize, number)))

print("\nОтражение относительно горизонтальной оси:")
printSquare(horizontalReflect(createColoredSquare(colorsAmount, squareSize, number)))

print("\nОтражение относительно главной диагонали:")
printSquare(diagonalReflect(createColoredSquare(colorsAmount, squareSize, number)))

print("\nОтражение относительно побочной диагонали:")
printSquare(subDiagonalReflect(createColoredSquare(colorsAmount, squareSize, number)))
```

### Решение задачи в общем виде

Обозначим за $n$ количество частей, на которые разбита сторона квадрата, а за $k$ - количество цветов в раскраске.

Как мы выяснили ранее $G = \{\tau,\,\tau^2,\,\tau^3,\,\sigma_1,\,\sigma_2,\,\sigma_3,\,\sigma_4,\,e\}$, $|G| = 8$.

#### Рассмотрим оба случая по отдельности:

1) $n$ - нечётное число:

В $\tau$ и в $\tau^3$ количество незивисимых циклов равно $\cfrac{n^2 - 1}{4} + 1$, так как длина каждого цикла, кроме одного, равна 4, и есть единственный центральный элемент, который остается на своем месте.

В $\tau^2$ количество независимых циклов равно $\cfrac{n^2 - 1}{2} + 1$, так как длина всех циклов, кроме одного, равна 2, и снова только центральный элемент остается на своем месте.

В $\sigma_1$ количество независимых циклов равно $\cfrac{n^2 - n}{2} + n$, так как длина каждого цикла, кроме n циклов, равна 2, а те самые n элементов на вертикальной оси остались на месте, ведь столбцу с ними не нашлось пары. Аналогично с $\sigma_2$.

---

2) $n$ - чётное число:

В $\tau$ и в $\tau^3$ количество незивисимых циклов равно $\cfrac{n^2}{4}$, так как длина каждого цикла равна 4, и ни один элемент не остается на своем месте.

В $\tau^2$ количество независимых циклов равно $\cfrac{n^2}{2}$, так как длина всех циклов, кроме одного, равна 2, и ни один элемент не остается на своем месте.

В $\sigma_1$ количество независимых циклов равно $\cfrac{n^2}{2}$, здесь отличие от случая с нечетным $n$ объясняется тем, что тому столбцу, которому не нашлось пары в прошлый раз, теперь пара нашлась. Аналогично с $\sigma_2$.

---

На количество циклов в $\sigma_3$ и $\sigma_4$ четность $n$ никак не влияет. В $\sigma_3$ и в $\sigma_4$ количество независимых циклов равно $\cfrac{n^2 - n}{2} + n$, так как длина каждого цикла, кроме n циклов, равна 2, а те самые n элементов на соответствующей диагонали остались на месте.

Теперь можно представить общую формулу:
$$
N = \begin{cases}
    \frac{1}{8}(k^{n^2}+2k^{\frac{n^2-1}{4} + 1}+k^{\frac{n^2-1}{2} + 1}+4k^{\frac{n^2-n}{2}+n}), \; if \; n \bmod 2 = 1 \\
    \frac{1}{8}(k^{n^2}+2k^{\frac{n^2}{4}}+3k^{\frac{n^2}{2}}+2k^{\frac{n^2-n}{2}+n}), \; else
\end{cases}
$$
