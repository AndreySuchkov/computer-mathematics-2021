## Основные теоретические положения

#### Определение 1

**Орбитой** $x \in X$ под действием $G$ называется множество $Gx = \{gx | g \in G\}$.

Тогда $|Gx|$ - **длина орбиты**, где  $G$ - группа с нейтральным элементом $e$.

#### Определение 2

Пусть $G$ - группа с нейтральным элементом $e$, а $X$ - произвольное множество. Тогда $G$ **действует** на X, если задана операция $\phi: G \times X \rightarrow X$. При этом пара $G \times X$ обозначается как $gx, g \in G, x \in X$.

Свойства операции $\phi$ $\forall x \in X, \forall g \in G$:

1. $ex = x$

2. $g(hx) = (gh)x$

### Лемма Бернсайда

Количество орбит N действия $G$ на $X$ задается следующей формулой: $N = \frac{1}{|G|}\sum_{g \in G} |X^g|$

#### Определение 3

Множество всех раскрасок  обозначим как $C^I$. Получаем $|C^I| = |C|^{|I|}$ , где $I$ - произвольное множество, а $C$ - множество цветов. $G$ - группа, действующая на множество $I$. Тогда формула $gf(i) = f(g^{-1}i)$, где $g \in G, i \in I, f \in C^I$, задает действие $G$ на множество $C^I$.

#### Лемма 1.

Пусть $g \in G, f \in C^I$. Тогда для этих $g$ и $f$ следующие условия эквивалентны:

1. $gf = f$

2. $f(i) = f(g^ni) \forall i \in I, \forall n \in Z$.

3. Если $i, j \in I$ лежат в одном цикле в циклической записи перестановки $\alpha(g)$, то $f(i) = f(g)$.


Таким образом , получаем , что количество раскрасок из $C^I$, которые сохраняют элемент $g \in G$ равно $|C|^{c(\alpha(g))}$, где $c(\alpha)$ - количество независимых циклов в перестановке $\alpha$.

## Реализация функции отображения раскраски:

```python
from sage.plot.polygon import polygon, polygon2d
import random

side = 3
height = side * pow(3,0.5)/2
colors = ['#B22222', '#A020F0', '#00BFFF', '#FFC125']

def make_triangle(x, y, color, reverse):
    edgecolor = 'black'
    p1 = (x, y)
    p2 = (x + side, y) if not reverse else (x - side/2, y + height)
    p3 = (x + side/2, y + height) if not reverse else (x + side/2, y + height)
    
    return polygon([p1, p2, p3], color=color, edgecolor=edgecolor)
    
    
def make_hex_half(start_x, start_y, y_direction, first_reverse, amounts, idt):
    hex_half = polygon([0, 0], axes=False)
    
    for amount in amounts:
        reverse = first_reverse
        x = start_x
        y = start_y
        
        for i in range(amount):
            color  = colors[idt % len(colors)]
            hex_half = hex_half + make_triangle(x, y, color, reverse)
            
            if not reverse:
                x = x + side

            reverse = not reverse
            idt = idt * 293 // 294
            
        start_x = start_x + side / 2
        start_y = start_y + y_direction * height
        
    return (hex_half, idt)

def make_hex(idt):
    hex_shape = polygon([0, 0], axes=False)
    triangles = [27, 25, 23, 21, 19, 17, 15]
    
    half, new_idt = make_hex_half(-27, 0, 1, False, triangles, idt)
    
    hex_shape = hex_shape + half + make_hex_half(-27 + side/2, -1 * height, -1, True, triangles, new_idt)[0]

    return hex_shape
```
