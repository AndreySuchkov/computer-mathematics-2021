# Основные теоретические положения

### Бинарный метод

Бинарный метод возведения в степень - это алгоритм, который позволяет возводить любое число x в степень n. В данном методе степень представляется в двоичной системе счисления. Далее происходит просмотр бит слева направо, не учитывая первый бит, так как он всегда равен единице: если бит равен единице, то текущее число возводится в квадрат и умножается на исходное чило, если бит равен нулю, то текущее число возводится в квадрат.

### Метод множителей

Метод множителей возведения числа в степень - это алгоритм, который позволяет возводить любое число x в степень n. Алгоритм следующий: если число n - простое, то x возводится в степень n-1, после чего домножается на x. Если же число n - составное, тогда степень числа представляется как произведение n = i*j, где i - наименьший простой множитель числа n. После этого число x возводится в степень i, а затем полученное число возводится в степень j.

### Аддитивная цепочка

Аддитивная цепочка для $n \in N$ - это последовательность $1=a_0, a_1, a_2, ... , a_m = n$, где $a_i = a_j + a_k, k \leq j \lt i, \forall i = \overline{1...m}$ . Длина минимальной аддитивной цепочки для числа $n$ равняется $l(n) = m$ Рекуррентная формула длины аддитивной цепочки: $l(n) = min\{l(n - 1) + 1, \ l_n\} + \sigma_n $

### Алгоритм Яо

Алгоритм Яо - это алгоритм построения аддитивной цепочки.

Выберем $n, k \geq 2$, тогда $n=\Sigma_{i=0}^j c_i2^{ik}, \quad c_i \ne 0$.

Функция $d(z) = \sum\limits_{i}2^{i*k}$, причем $\forall i: c_i = z$.

Такая аддитивная цепочка будет начинаться с последовательности степеней двойки, не превышающих числа $n$: $Y_k(n) = 1, 2, 4,..., 2^{\lambda(n)}$, далее необходимо вычислить все $d(z)$: $\forall z \in \{1,2,3,...,2^k-1\}, d(z) \neq 0$

Теперь $n$ можно представить в виде $n=\Sigma_{z=1}^{2^{k-1}}zd(z)$ (происходит перевод в систему счисления с основанием $2^k$). Результаты дописываются в аддитивную цепочку.

### Звёздная цепочка

Аддитивная цепочка называется звездной, если она включает в себя только звездные шаги $a_i = a_{i-1} + a_k, k = \overline {0...(i-1)}$.

### Алгоритм дробления вектора индексов

Для данного $n \in N$ надо найти $\min l^*(n) = m: a_m = n$ 

Алгоритм: 
1) Во внешнем цикле рассматриваются аддитивные цепочки длины $m$: $\underline{l}(n) \leq m \leq \overline{l}(n)$, где $1 \leq q \leq m - 1$, пусть $q = \frac{m}{2}$, $q \in N$

2) Во внутреннем цикле перебираются все фиксированные части вектора индексов $\{r_i\}_{i=1}^{q}$ ($q!$ шагов) на каждом шаге строим $\{r_i\} \to \{a_i\}$, вычисляем $a_{min}$ и $a_{max}$

-если на каком-то шаге $a_m = n$, то задача решена.

-если $n \notin [a_{min}, a_{max}]$, то переходим к следующему набору $\{r_i\}_{i=1}^{q}$

-если $n \in [a_{min}, a_{max}]$, то организуем внутренний цикл перебора меняющейся части $\{\rho_i\}_{q+1}^{m}$

-если $\rho_i = 1,при \ \forall j$, то переходим к следующему набору $\{r_i\}_{i=1}^{q}$

3) Если $r_i = 1,при \ \forall i$ и $\rho_i = 1, при \ \forall j$ ,то $m = m + 1$

### Теорема Брауэра

Для числа $k$&lt; $log_2 (log_2 n)$ справедливо, что: $l(n)$&lt; $(1+k^{-1}) \cdot \lceil log_2 n \rceil + 2^{k-1} - k + 2$. Если $k=\lambda(\lambda(n))-2\lambda(\lambda(\lambda(n)))$, то следует что:

Следствие 1: $\lim \limits_{n \to \infty} \frac{l(n)}{\lambda(n)} = 1$

Следствие 2: Одна из наилучших верхних оценок длины аддитивной цепочки имеет вид: $\lambda(n)(1 + \frac{1}{\lambda(\lambda(n))} + \frac{o(\lambda(\lambda(\lambda(n))))}{\lambda(\lambda(n)) ^ 2})$

### Гипотеза Шольца-Брауэра

$l^*(2^n-1) \leq l^*(n)+n-1$

## 1. Бинарный метод и метод множителей

### Бинарный метод

```python
def BinaryMethod(x, n):
    step = 0
    res = x
    listDeg = [1]
    temp = bin(n)[3:]
    for i in temp:
        if i == '1':
            listDeg.append(listDeg[-1] + listDeg[-1])
            res *= res
            listDeg.append(listDeg[-1] + 1)
            res *= x
            step += 2
        else:
            listDeg.append(listDeg[-1] + listDeg[-1])
            res *= res
            step += 1
    return res, step, listDeg
```

### Метод множителей

```python
def MinPrime(n):
    if n % 2 == 0:
        return 2
    d = 3
    while d * d <= n and n % d != 0:
        d += 2
    if d * d <= n:
        return d
    return n

def MulMethod(x, n):
    minDiv = MinPrime(n)
    x1, step1, listDeg1 = BinaryMethod(x, minDiv)
    x2, step2, listDeg2 = BinaryMethod(x1, n/minDiv)
    return step1+step2, listDeg1, listDeg2
```

### Сравнение бинарного метода и метода множителей

```python
from matplotlib import pyplot as plt
resTable = []
x = 1
n = [5, 17, 29, 33, 44, 71, 137, 338, 445]
for i in n:
    res, stepBin, listDegB = BinaryMethod(x, i)
    stepMul, listDegM1, listDegM2 = MulMethod(x, i)
    resTable.append([i, stepBin, stepMul])
    print("Последовательность степеней для бинарного метода: ", listDegB)
    print("Последовательность степеней для метода множителей: x:", listDegM1, "y = x ^", listDegM1[-1],": ", listDegM2)
    print()

headers = ["Степень", "Количество шагов бинарного метода", "Количество шагов метода множителей"]
cell = []
for i in resTable:
    cell.append([data for data in i])

table = plt.table(cell, rowLoc='right', colLabels=headers, loc='center')
table.set_fontsize(20)
table.scale(4, 2) 

ax = plt.gca()
ax.get_xaxis().set_visible(False)
ax.get_yaxis().set_visible(False)

plt.box(on=None)

plt.draw()
```

## 2. Алгоритм Яо

```python
def Dz(d, z):
    listDeg = []
    listDeg.append(d)
    while listDeg[len(listDeg)-1] != d*z:
        flag = false
        i = 1
        while(flag == false):
            if listDeg[len(listDeg)-i] + listDeg[len(listDeg)-1] <= z*d:
                listDeg.append(listDeg[len(listDeg)-i] + listDeg[len(listDeg)-1])
                flag = true
            else:
                i += 1
    return listDeg

def AlgorithmYao(n, k):
    listDeg = []
    i = 1
    while i < n:
        listDeg.append(i)
        i = i << 1
    
    digits = n.digits(1 << k)
    
    j = 1
    last = []
    while j <= ((1 << k) - 1):
        d = 0
        for m in range(len(digits)):
            if digits[m] == j:
                d += (1 << k*m)
        if d != 0:
            listDeg += Dz(d, j)
            last.append(listDeg[len(listDeg)-1])
        j += 1
    listDeg.append(sum(last))
    return listDeg

def MinLen(n):
    return sum(map(int, bin(n)[2:])) + len(bin(n)[2:]) - 1

from matplotlib import pyplot as plt
resTable = []
n = [28, 57, 93, 148, 473]
k = [2,3,4,5]
for i in n:
    for j in k:
        listDeg = AlgorithmYao(i, j)
        resTable.append([i, j, listDeg, len(listDeg), MinLen(i)])

headers = ["Степень", "Параметр k", "Аддитивная цепочка", "Длина цепочки(Яо)", "Минимальная длина цепочки"]
cell = []
for i in resTable:
    cell.append([data for data in i])

table = plt.table(cell, rowLoc='right', colLabels=headers, loc='center')
table.set_fontsize(20)
table.scale(4, 2) 
table.auto_set_column_width(col=list(range(3)))
ax = plt.gca()
ax.get_xaxis().set_visible(False)
ax.get_yaxis().set_visible(False)

plt.box(on=None)

plt.draw()
```

## 3. Алгоритм дробления вектора индексов

```python
from matplotlib import pyplot as plt
import time
import math



def StarChain(vi):#функция для создания звёздной цепочки
    chain = [1]
    for i in vi:
        chain.append(chain[-1] + chain[i - 1])
    return chain


def NextSeq(sequence, q):#функция для нахождения следующего набора
    if (sequence == [1 for i in range(len(sequence))]):
        return [0 for i in range(len(sequence))]
    for i in range(len(sequence) - 1, -1, -1):
        if (sequence[i] == 1):
            sequence[i] = q + i + 1
            continue
        sequence[i] -= 1
        break
    return sequence


def VectorSplittingAlgorithm(n):#наш алгоритм
    for m in range(int(math.log(n, 2)), int(math.log(n, 2)) + bin(n)[3:].count('1') + 1):
        q = m // 2
        seqR = [i for i in range(1, q + 1)]
        while (seqR != [0 for i in range(q)]):
            seqRo = [q + i + 1 for i in range(m - q)]
            chain = StarChain(seqR + seqRo)
            aMin = chain[q] + (m - q)
            aMax = chain[q] * 2 ** (m - q)
            if (n < aMin or n > aMax):#если n не входи в промежуток
                seqR = NextSeq(seqR, 0)#переход к следующему набору r
                continue
            if(chain[-1] == n):
                return chain
            while (seqRo != [0 for i in range(m - q)]):
                chain = StarChain(seqR + seqRo)
                if (n == chain[-1]):
                    return chain
                seqRo = NextSeq(seqRo, q)#переход к следующему набору р
            seqR = NextSeq(seqR, 0)#переход к следующему набору r
    return

resTable = []
n = [513, 547, 618, 710]
for i in n:
    begin = time.time()
    chain = VectorSplittingAlgorithm(i)
    end = time.time()
    resTable.append([i, chain, len(chain), end-begin])

headers = ["Степень", "Звёздная цепочка", "Длина цепочки","Потрачено времени (сек)"]
cell = []
for i in resTable:
    cell.append([data for data in i])

table = plt.table(cell, rowLoc='right', colLabels=headers, loc='center')
table.set_fontsize(20)
table.scale(4, 2) 
table.auto_set_column_width(col=list(range(3)))
ax = plt.gca()
ax.get_xaxis().set_visible(False)
ax.get_yaxis().set_visible(False)

plt.box(on=None)

plt.draw()
```

```python
import pylab as plt
import math

def StarChain(vi):#функция для создания звёздной цепочки
    chain = [1]
    for i in vi:
        chain.append(chain[-1] + chain[i - 1])
    return chain


def NextSeq(sequence, q):#функция для нахождения следующего набора
    if (sequence == [1 for i in range(len(sequence))]):
        return [0 for i in range(len(sequence))]
    for i in range(len(sequence) - 1, -1, -1):
        if (sequence[i] == 1):
            sequence[i] = q + i + 1
            continue
        sequence[i] -= 1
        break
    return sequence


def VectorSplittingAlgorithm(n):#алгоритм дробления вектора индексов
    for m in range(int(math.log(n, 2)), int(math.log(n, 2)) + bin(n)[3:].count('1') + 1):
        q = m // 2
        if (q == 0):
            q = 1
        seqR = [i for i in range(1, q + 1)]
        while (seqR != [0 for i in range(q)]):
            seqRo = [q + i + 1 for i in range(m - q)]
            chain = StarChain(seqR + seqRo)
            aMin = chain[q] + (m - q)
            aMax = chain[q] * 2 ** (m - q)
            if (n < aMin or n > aMax):#если n не входи в промежуток
                seqR = NextSeq(seqR, 0)#переход к следующему набору r
                continue
            if(chain[-1] == n):
                return chain
            while (seqRo != [0 for i in range(m - q)]):
                chain = StarChain(seqR + seqRo)
                if (n == chain[-1]):
                    return chain
                seqRo = NextSeq(seqRo, q)#переход к следующему набору р
            seqR = NextSeq(seqR, 0)#переход к следующему набору r
    return []

def TheoremeBrauer(n):
    l_n = [len(VectorSplittingAlgorithm(i)) for i in range(1, n)]
    x = [i for i in range(1, n)]
    x_ln = [math.log(i, math.e) for i in range(1, n)]
    x_ln2 = [math.log(i, math.e)**2 for i in range(1, n)]
    y_i = [l_n[i-1] * x_ln[i-1] for i in range(1, n)]
    f = squaresMethod(n, y_i, x_ln, x_ln2, l_n)
    return x, f, x_ln, l_n

def squaresMethod(n, y_i, x_ln, x_ln2, l_n):
    b = (n*sum(y_i) - sum(l_n)*sum(x_ln))/(n*sum(x_ln2) - sum(x_ln)*sum(x_ln))
    a = (sum(l_n) - b * sum(x_ln))/(n-1)
    f = [a + b*x_ln[i-1] for i in range(1, n)]
    return f
 
    
n = 201
x, y, x_ln, l_n = TheoremeBrauer(n)


plt.figure(figsize=(10, 10))
plt.scatter(x, l_n, color = 'red', marker = '.')
plt.plot(x, y, color = 'blue')
plt.show()
```

## 5. Гипотеза Шольца-Брауэра

```python
import pylab as plt
import math

def StarChain(vi):#функция для создания звёздной цепочки
    chain = [1]
    for i in vi:
        chain.append(chain[-1] + chain[i - 1])
    return chain


def NextSeq(sequence, q):#функция для нахождения следующего набора
    if (sequence == [1 for i in range(len(sequence))]):
        return [0 for i in range(len(sequence))]
    for i in range(len(sequence) - 1, -1, -1):
        if (sequence[i] == 1):
            sequence[i] = q + i + 1
            continue
        sequence[i] -= 1
        break
    return sequence


def VectorSplittingAlgorithm(n):#алгоритм дробления вектора индексов
    for m in range(int(math.log(n, 2)), int(math.log(n, 2)) + bin(n)[3:].count('1') + 1):
        q = m // 2#т.к. m маленькое, то q может быть 0.
        if (q == 0):#добавляем условие, при котором q становится 1
            q = 1
        seqR = [i for i in range(1, q + 1)]
        while (seqR != [0 for i in range(q)]):
            seqRo = [q + i + 1 for i in range(m - q)]
            chain = StarChain(seqR + seqRo)
            aMin = chain[q] + (m - q)
            aMax = chain[q] * 2 ** (m - q)
            if (n < aMin or n > aMax):#если n не входи в промежуток
                seqR = NextSeq(seqR, 0)#переход к следующему набору r
                continue
            if(chain[-1] == n):
                return chain
            while (seqRo != [0 for i in range(m - q)]):
                chain = StarChain(seqR + seqRo)
                if (n == chain[-1]):
                    return chain
                seqRo = NextSeq(seqRo, q)#переход к следующему набору р
            seqR = NextSeq(seqR, 0)#переход к следующему набору r
    return []

resTable = []
for i in range(2,11): #до 11
        seq1 = VectorSplittingAlgorithm((1 << i) - 1)
        seq2 = VectorSplittingAlgorithm(i)
        resTable.append([i, len(seq1), len(seq2) + i - 1])
        
headers = ["Степень", "Длина аддитивной цепочки (алгоритм дробления вектора индексов)", "Длина цепочки по гипотезе Шольца-Брауэра"]
cell = []
for i in resTable:
    cell.append([data for data in i])

table = plt.table(cell, rowLoc='right', colLabels=headers, loc='center')
table.set_fontsize(20)
table.scale(4, 2) 
table.auto_set_column_width(col=list(range(3)))
ax = plt.gca()
ax.get_xaxis().set_visible(False)
ax.get_yaxis().set_visible(False)

plt.box(on=None)

plt.draw()
```

Из таблицы можно сделать вывод, что для чисел $n \leq 10$ гипотеза справедлива. Следовательно, если необходимо вычислить длину аддитивной цепочки для этих чисел, то можно воспользоваться гипотезой и вычислять длину по формуле $l^*(n)+n-1$.
