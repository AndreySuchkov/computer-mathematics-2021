## Основные теоретические положения 

Пусть $G$ - ***группа с нейтральным элементом*** $e$, а $X$ - множество. $G$ действует на $X$, если задана операция $G \times X \to X$, обладающая для $\forall x \in X$ и $\forall g,h \in G$ следующими свойствами:

1. $ex = x$,
2. $g(hx) = (gh)x$,

где $gx$ - это образ пары $(g,x)$.

***Орбитой элемента*** $x \in X$ под действием $G$ называется множество $Gx = \{gx | g \in G\}$. Количество элементов в орбите - длина орбиты. Любые две орбиты либо не пересекаются, либо совпадают. Таким образом, множество $X$ разбивается в дизъюнктивное объединение орбит.

***Неподвижными точками*** элемента $g \in G$ называются те $x \in X$, для которых $gx = x$. Множество неподвижных точек элемента $g$ обозначается через $X^{g}$.

По ***лемме Бёрнсайда*** количество орбит действия группы ${G}$ на множество ${X}$ равно: 

$${N = \frac{1}{\left | G \right |}\sum_{g\in G}(\left| X^g \right|)}.$$

Пусть $I$ - произвольное множество, а $C$ - множество цветов. ***Расскраской множества*** $I$ называется функция из $I$ в $C$. Множество всех раскрасок обозначается через $C^I$. $|C^I| = |C|^{|I|}$. Пусть $G$ группа, действующая на множестве $I$, тогда формула $gf(i) = f(g^{-1}i)$, где $g \in G$, $i \in I$, а $f \in C^{I}$ задает действие $G$ на $C^{I}$. А для элемента $g \in G$ и функции $f \in C^{I}$ следующие условия эквивалентны:

1. $gf = f$,
2. $f(i) = f(g^{n}i)$, $\forall i \in I $ и $n \in \mathbb{Z}$,
3. Если $i,j \in I$ лежат в одном цикле в циклической записи перестановки $\varphi(g)$, то $f(i) = f(j)$.

***Количество раскрасок*** из $C^{I}$, которые сохраняет данный элемент $g \in G$ равно $|C|^{c(\varphi(g))}$, где $c(\varphi(g))$ - количество независимых циклов перестановки $\varphi(g)$ (считая циклы длины 1).
