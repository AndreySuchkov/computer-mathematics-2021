Пусть $\mathbb{G}$ - группа с нейтральным элементом $e$, $\mathbb{X}$ - множество.

$\mathbb{G}$ действует на $\mathbb{X}$, если $\mathbb{G} \times \mathbb{X} \rightarrow \mathbb{X}$ такая, что $\forall x \in \mathbb{X}$, $\forall g, h \in \mathbb{G}$ выполняется:

1. $e x = x$
2. $g (h x) = (g h) x$

**Орбита:**

Орбитой $x \in \mathbb{X}$ под действием группы $\mathbb{G}$ называется:

$Gx = \{gx| g \in \mathbb{G}\} = Orb(x)$.

**Неподвижные точки:**

Неподвижные точки элемента $g \in \mathbb{G}$ - это такие $x \in \mathbb{X}$, что $gx = x$. Обозначение множества неподвижных точек: $\mathbb{X}^g$.

**Стабилизатор:**

Стабилизатор - это множество всех преобразований из $\mathbb{G}$, которые оставляют $x$ на месте.

$St(x) = \{g \in \mathbb{G}| gx = x\}$, $St \subset \mathbb{G}$.

Длина орбиты равна индексу стабилизатора. 

**NB!** $\#(\mathbb{G}x)*\#St(x) = \#\mathbb{G}$.

**Лемма Бернсайда:**

Количество орбит действия группы $\mathbb{G}$ на множество $\mathbb{X}$ равно:

$\frac{1}{\#\mathbb{G}}\sum_{g\in \mathbb{G}}(\#\mathbb{X}^g)$


**Раскраски:**

Пусть $I$ - произвольное множество, $C$ - множество цветов.

Раскраской множества $I$ называется функция из $I$ в $C$. Где $C^I$ - множество всех раскрасок.

**Действие $\mathbb{G}$ на $C^I$:**

Путь $\mathbb{G}$ - группа, действующая на $I$, тогда $gf(i) = f(g^{-1}(i))$, где $g \in \mathbb{G}$, $i \in I$, $f \in C^I$.

**Теорема:** 

$g \in  \mathbb{G}$, $f \in C^I$.

1. $gf = f$
2. $f(i) = f(g^n(i))$, $\forall i \in I$, $n \in \mathbb{Z}$
3. Если $i, j \in I$ лежат в одном цикле в циклической записи $\alpha(g)$, то $f(i) = f(j)$.

**Следствие:**

Количество раскрасок из $C^I$, которые сохраняют данный элемент $g \in \mathbb{G} = (\#C)^{c(\alpha(g))}$, где $c(\alpha)$ - количество независимых циклов в перестановке $\alpha$.

Раскраски, совмещаемые друг с другом каким-нибудь преобразованием из группы $\mathbb{G}$, будут задавать одинаковый круг.
