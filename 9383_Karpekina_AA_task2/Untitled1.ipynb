1) $ex = x$

Определение неподвижных точек.
Неподвижными точками элемента $g \in G$ называются такие $x \in X$, для которых $gx = x$. Множество неподвижных точек элемента $g$ обозначается через $X^{g}$.

Лемма Бёрнсайда.
Количество орбит действия группы $G$ на множестве $X$ равно: $\frac{1}{|G|}\sum\limits_{g\in G}|X^{g}|$

```python
import matplotlib
from matplotlib.patches import Wedge
import matplotlib.pyplot as plt
```

```python
def printCircle(temp):
    fig=plt.figure(figsize=(12, 12))
    ax=fig.add_subplot(111) 
    plt.axis([-1, 1, -1, 1])
    color_dict = {1 : 'red', 2 : 'green', 3 : 'blue', 4 : 'yellow'}

    dx=360/len(temp)
    print(dx)
    for i in range(0,len(temp)):
        fov1 = Wedge((0,0), 1, i*dx, i*dx+0.1,color='black')
        fov2 = Wedge((0,0), 1, i*dx+dx-0.1, i*dx+dx,color='black')

        fov = Wedge((0,0), 1, i*dx, i*dx+dx,color=color_dict[temp[i][1]])
        ax.add_artist(fov)
        ax.add_artist(fov1)
        ax.add_artist(fov2)

        

    plt.show()
```
