## Основные теоретические положения.
### Бинарный метод.
Бинарный метод - алгоритм быстрого возведения числа в натуральную степень. 
В начале работы алгоритма натуральная степень представляется в бинарном виде(переводится в двоичную степень счисления), после из бинарной записи числа удаляется старший бит. Затем бинарная запись анализируется и производится условная "замена" каждой 1 на "КУ", каждого 0 на "К", где каждое К - умножение числа само на себя, а У - умножение числа на X. Затем производится само возведение в степень с помощью базовых операций умножения.
### Метод множителей.
Метод множителей - алгоритм быстрого возведения числа в натуральную степень. 
В начале работы алгоритма производится анализ показателя степени k. Если k - простое число, то производится работа с
числом (k-1), в конце работы алгоритма производится домножение на число x. Далее k представляется в виде произведения $k=ij$, где i - наименьший простой множитель k. Теперь производится возведение $x^i$ бинарным методом. Затем производится замена $x^i = y$, и производится возведение $y^j$ бинарным методом и производится обратная замена.
Данный метод для некоторых чисел дает более быстрый результат возведения, чем бинарный метод.
### Аддитивные цепочки.
Аддитивной цепочкой для $n \in N$ называется последовательность $1=a_0,a_1,a_2,...,a_m=n$, обладающих свойством: $a_i=a_j+a_k, где k \leq j \leq i \forall i=1..m$. Исходя из определения получаем, что $a_i=a_j+a_k$ - шаг аддитивной цепочки.

Типы шагов:

1. Если $j=k=i-1$, то i-ый шаг АЦ есть удвоение
2. Если $j=i-1, k \in [0, i-1]$, то i-ый шаг есть звездный
3. Если $\lambda(a_i)=\lambda(a_{i-1})$, то i-ый шаг есть малый шаг

Дадим определение наименьшей длины аддитивной цепочки при $n\leq 1000$:

$l(n)$ - наименьшая длина аддитивной цепочки

При $n\leq 1000$ $l(n) = min\{l(n-1)+1, l_n\}-\sigma_n$, где $l_p = \begin{cases} \infty,\;если\,p - простое \\ l(q)+l(\frac{p}{q}),\; где\,q - минимальный\,простой\,делитель\,p \\ \end{cases}$ 

$\sigma_n = \{0,1\}$
### Алгоритм Яо.
Алгоритм Яо - алгоритм построения примерной аддитивной цепочки длины n. Цепочка, которая получается в результате работы алгоритма еще называют цепочкой Яо. Она состоит из двух цепочек. Первая цепочка представляет собой последовательность $Y_k(n): 1, 2,4,...,2^{\lambda(n)}.$ Вторая цепочка уже строится на основе первой. Подробности построения указаны в описании работы алгоритма.

В начале работы алгоритма имеем $n,k\geq 2$, где $n=\sum_{i=0}^j{a_i2^{ik}}, a_i \neq 0$, где $a_i$ - цифра i-го разряда числа, $j = 2^k-1$. Затем ведется подсчет $zd(z)$ для всех чисел от 1 до j. Получают $zd(z)$ с помощью сложения чисел из $Y_k(n)$. Затем с помощью сложений всех $zd(z)$ отличных от нуля получается число $n$. $zd(z)$ и представляет собой вторую цепочку Яо. 
### Алгоритм дробления вектора индексов.
#### Звездные цепочки.
Звездные цепочки - цепочки, состоящие из звездных шагов. Звездный шаг представляет собой сумму предыдущего шага и какого-либо еще, т.е $a_i=a_{i-1}+a_k$, где $0\leq k \leq i-1$
#### Вектор индексов.
Вектор индексов представляет собой последовательность натуральных чисел вида $\{r_i\}_{i=1}^{m-1}$, где $r_i = \{x: 1\leq x \leq i\}$. Формула для вычисления элементов звездной цепочки по вектору индексов $a_i = a_{i-1}+a_{r_{i-1}}$, где $2 \leq i \leq m-1$
#### Алгоритм.
Алгоритм представляет собой перебор векторов, но с улучшениями, которые позволяют быстрее прийти к результату, чем при простом переборе. Алгоритм заключается в разделении вектора на два других: статичный и основной, который будет меняться. Длина внешнего цикла $m$ находится следующим образом $\underline{l}(n) \leq m \leq \overline{l}(n)$, где $\underline{l}(n) = \lceil lbn\rceil, \overline{l}(n)=\lambda(n)+\nu(n)-1$. При этом длина внутреннего цикла находится в пределах $1 \leq q \leq m-1$. Далее считаются $a_{min}, a_{max}$. Далее на каждом шаге строится звездная цепочка. Если $a_m \in [a_{min}, a_{max}]$, то организуется перебор внутреннего цикла, в другом случае происходит переход на новый цикл. Если $a_m = n$, то звездная цепочка найдена и алгоритм прекращает свою работу.
### Следствие 1 теоремы Брауэра.
$\lim\limits_{n \to \infty}\frac{l(n)}{\lambda(n)}=1$
### Гипотеза Шольца-Брауэра.
Для зведной цепочки, при $n \leq 64$ верно $l^*(2_n-1) \leq l^*(n)+n-1$, где $l^*(n)$ - длина минимальной аддитивной цепочки для числа $n$

## 1. Решение задач вычисления степеней бинарным методом и методом множителей.

```python
def binaryMethod(x, deg):
    numberOfOperation = 0
    y = x
    a = bin(deg)
    a = a[3:len(a)]
    for i in a:
        if i == '1':
            x *= x
            numberOfOperation += 1
            x *= y
            numberOfOperation += 1
        else:
            x *= x
            numberOfOperation += 1
    L = []
    L.append(x)
    L.append(numberOfOperation)
    return L


def checkEasy(a):
    counter = 1
    for i in range(2, a-1):
        if a % i == 0:
            counter = 0
            break
    return counter


def findEasy(a):
    for i in range(2, a-1):
        if a % i == 0:
            return i
    return a


def multMethod(x, deg):
    flag = checkEasy(deg)
    if flag == 1:
        deg -= 1
    i = findEasy(deg)
    j = int(deg / i)
    L = binaryMethod(x, i)
    y = L[0]
    k = L[1]
    L = binaryMethod(y, j)
    L[1] += k
    if flag == 1:
        L[0] *= x
        L[1] += 1
    return L
    

L = []
numList = [43, 77]
for i in numList:
    line = []
    a = binaryMethod(2, i)
    line = [str(i), a[1]]
    a = multMethod(2, i)
    line.append(a[1])
    L.append(line)
    
figure, ax = plt.subplots()

header = ['n', 'Бинарный метод', 'Метод множителей']
ax.set_axis_off()
table = ax.table(cellText = L, colLabels = header, cellLoc = 'center', loc = 'upper left')
table.auto_set_font_size(False)
cellDict = table.get_celld()
for i in range(len(numList)+1):
    cellDict[(i,0)].set_width(0.2)
    cellDict[(i,1)].set_width(0.7)
    cellDict[(i,2)].set_width(0.4)
table.set_fontsize(14)
table.scale(2,3)
```

## 2. Реализация алгоритма Яо

```python
def yao(n, k):
    output = []
    counter  = -1
    for i in range(0, math.floor(math.log2(n))+1):
        output.append(2**i)
        counter += 1
    base = 2**k
    decompose = []
    index = 0
    while n != 0:
        digit = index*k
        decompose.append((n % base, digit))
        n //= base
        index += 1
    decompose.reverse()
    work_dict = {}
    yao_chain = []
    while decompose:
        a = decompose.pop()
        z, deg = a
        deg = int(math.pow(2, deg))
        temp_turple = (z, deg)
        if z not in work_dict:
            if z == 0:
                continue
            work_dict.update([temp_turple])
        else:
            work_dict[z] += deg
    for i in work_dict.values():
        yao_chain.append(i)
        counter += 1
    check_n = 0
    for i in work_dict.keys():
        temp = work_dict.get(i)
        a = bin(i)
        a = a[3:len(a)]
        for j in a:
            if j == '1':
                temp += temp
                yao_chain.append(temp)
                counter += 1
                temp += work_dict.get(i)
                yao_chain.append(temp)
                counter += 1
            else:
                temp += temp
                yao_chain.append(temp)
                counter += 1
        check_n += temp
    yao_chain.append(check_n)
    counter += 1
    yao_chain.sort()
    output.extend(yao_chain)
    return output, counter


L = []
numList = [553, 602]
minLenght = {'553':'11', '602':'12'}
for i in numList:
    for j in range(5)[2:]:
        line = []
        chain, a = yao(i,j)
        line = [str(i),str(j),len(chain),str(minLenght.get(str(i))),str(chain)]
        L.append(line)
    
figure, ax = plt.subplots(1,1)

header = ['n', 'k', 'Длина АЦ', 'Минимальная длина АЦ', 'Аддитивная цепочка']
ax.set_axis_off()
table = ax.table(cellText = L, colLabels = header, cellLoc = 'center', loc = 'upper left')

table.auto_set_font_size(False)
cellDict = table.get_celld()
for i in range(len(numList)*3+1):
    cellDict[(i,0)].set_width(0.1)
    cellDict[(i,1)].set_width(0.1)
    cellDict[(i,2)].set_width(0.2)
    cellDict[(i,3)].set_width(0.5)
    cellDict[(i,4)].set_width(1.2)
    
table.set_fontsize(14)
table.scale(2,3)

plt.show()
```

## 3. Реализация метода дробления индексов

```python
import math
import time
import matplotlib.pyplot as plt

def L_n(n):
    return int(math.log2(n))


def L_N(n):
    a = int(math.log2(n))
    b = 0
    for i in bin(n):
        if i == '1':
            b += 1
    return a + b - 1


def getChain(vector):
    chain = [1]
    for index in vector:
        chain.append(chain[index-1] + chain[-1])
    return chain


def decVec(vec, q, m):
    maxVec = [a for a in range(q,m+1)]
    minVec = [1 for a in range(q,m+1)]
    zeroVec = [0 for a in range(q,m+1)]
    if vec == minVec:
        return zeroVec
    ind = 1
    while (vec[-ind] == 1) and (ind < len(vec)):
        vec[-ind] = maxVec[-ind]
        ind += 1
    vec[-ind] -= 1
    return vec


def lamb(number):
    return int(math.log(number, 2) // 1)


def nu(number):
    count = 0
    while number != 0:
        count += number % 2
        number = number // 2
    return count


def getLowerBound(number):
    addition = 0
    if number % 1 != 0:
        addition = 1
    return (lamb(number) + addition)


def getUpperBound(number):
    return (nu(number) + lamb(number) - 1)

def indexMethod(n):
    l_n = L_n(n)
    l_N = L_N(n)
    for m in range(l_n, l_N+1):
        halfChain = []
        q = m//2
        beginVector = [a for a in range(1, q+1)]
        endVector = [0 for a in range(q)]
        while beginVector != endVector:
            halfChain = getChain(beginVector)
            a_max = halfChain[-1] * 2**(m-q)
            a_min = halfChain[-1] + m - q
            beginVector1 = [a for a in range(q+1, m+1)]
            endVector1 = [0 for a in range(m-q)]
            starChain = []
            while beginVector1 != endVector1:
                if a_min > n or a_max < n:
                    break
                starChain = getChain(beginVector + beginVector1)
                if starChain[-1] == n:
                    return starChain
                beginVector1 = decVec(beginVector1, q+1, m)
            beginVector = decVec(beginVector, 1, q)


L = []
numList = [553, 602]
for i in numList:
    t_beg = time.time()
    starChain = indexMethod(i)
    t_end = time.time()
    line = []
    line = [str(i), str(starChain), str(t_end-t_beg)]
    L.append(line)
    
figure, ax = plt.subplots()

header = ['Число', 'Аддитивная цепочка', 'Время, с']
ax.set_axis_off()
table = ax.table(cellText = L, colLabels = header, cellLoc = 'center', loc = 'upper left')
table.auto_set_font_size(False)
cellDict = table.get_celld()
for i in range(len(numList)+1):
    cellDict[(i,0)].set_width(0.2)
    cellDict[(i,1)].set_width(0.7)
    cellDict[(i,2)].set_width(0.4)
table.set_fontsize(14)
table.scale(2,3)

plt.show()
```

Из таблицы видно, что метод дробления индексов позволяет найди минимальную звездную цепочку. Но данный алгоритм обладает достаточно большим временем выполнения, что является следствием того, что алгоритм представляет собой перебор.

## 4. Проверка следствия Брауэра

```python
import matplotlib.pyplot as _pyplot
import numpy as _numpy
from scipy.optimize import leastsq

def residuals(coeffs, y, t):
    return  y - _numpy.log(t)*coeffs[0]

x = []
y = []
for i in range(1,201):
    x.append(i)
    y.append(len(str(indexMethod(i))))

figure, axes = _pyplot.subplots(figsize=(10,10))
p, flag = leastsq(residuals, 1, args=(y, x))
_pyplot.plot(x,y,'o',x, _numpy.log(x)*p[0], color='green')
_pyplot.show()
```

Функция $l(n)$, как видно из графика, аппроксимируется логарифмической кривой. Можно заключить, что при устремлении n к бесконечности, функция будет незначительно отличаться от логарифмической кривой. Это подтверждает следствие Брауэра.

## 5. Проверка гипотезы Шольца-Брауэра

```python
x = list(range(1,11))
starChain1 = []
starChain2 = []
for i in range(1,11):
    starChain1.append(len(str(indexMethod(2**i-1))))
    starChain2.append(len(str(indexMethod(i))) + i -1)
fig, ax = _pyplot.subplots(figsize=(10,10))
ax.plot(x,starChain1,'red',label="l(n^2-1)")
ax.plot(x,starChain2,'blue',label="l(n)+n-1")
ax.legend()
ax.grid()
_pyplot.show()
```

Исходя из построенных графиков, можно заключить, что гипотеза Шольца-Брауэра верна, так как $l(n^2-1) \leq l(n)+n-1$
