## Основные теоретические положения
**Действие группы на множество** -

Пусть $G$ - группа с нейтральным элементом $e$, а $X$ - множество. Будем говорить, что $G$ действует на $X$, если задана операция $G \times X \to X$, обладающая для любого $x \in X$ и $g, h \in G$ свойствами $ex = x$ и $g(hx) = (gh)x$

**Симметрическая группа** - множество всех биективных функций $X \to X$ с операцией композиции на множестве $X$. Обозначается $S_X$

**Орбита элемента $x \in X$ под действием $G$** - множество $Gx = \{gx | g \in G\}$

**Множество неподвижных точек элемента $g$** - множество, состоящее из $x \in X$ таких, что $gx = x$. Обозначается через $X^g$

**Лемма Бернсайда** - количество орбит действия группы $G$ на множество $X$ равно
$\cfrac{1}{|G|}\sum_{g\in G}|X^g|$

```python
d1d2 = Permutation(((1,2),(7,8), (4, 6), (3, 5)))
d1d3 = Permutation(((1,5),(3,7), (4, 6), (2, 8)))
d1d4 = Permutation(((1,4),(7,6), (2, 8), (3, 5)))
d2d3 = Permutation(((2,3),(5,8), (4, 6), (1, 7)))
d2d4 = Permutation(((2,6),(4,8), (3, 5), (1, 7)))
d3d4 = Permutation(((3,4),(5,6), (2, 8), (1, 7)))
e = Permutation([1,2,3,4,5,6,7,8])
permutations_d = {'e':e, '(d1d2)':d1d2,'(d1d3)':d1d3, '(d1d4)':d1d4, '(d2d3)':d2d3, '(d2d4)': d2d4, '(d3d4)':d3d4} 
permutations_d['(d1d2)(d3d4)'] = d1d2*d3d4
permutations_d['(d1d3)(d2d4)'] = d1d3*d2d4
permutations_d['(d1d4)(d2d3)'] = d1d4*d2d3
three = Permutations([1,2,3, 4], 3)
three = three.list()
done = []
del three[-12:]
for e in three:
    d1_index = [e[0], e[1]]
    d2_index = [e[1], e[2]]
    d1_index.sort()
    d2_index.sort()
    a0 = '(d' + str(e[0]) + 'd' + str(e[1]) +  'd' + str(e[2]) + ')'
    a1 = '(d' + str(d1_index[0]) + 'd' + str(d1_index[1]) + ')'
    a2 = '(d' + str(d2_index[0]) + 'd' + str(d2_index[1]) + ')'

    res = permutations_d[a1] * permutations_d[a2]

    if res not in done:
        done.append(res)
    else:
        continue
    permutations_d[a0] = res
four = Permutations([1,2,3, 4], 4).list()
del four[-18:]

for e in four:
    d1_index = [e[0], e[1], e[2]]
    d2_index = [e[2], e[3]]
    d2_index.sort()
    a0 = '(d' + str(e[0]) + 'd' + str(e[1]) +  'd' + str(e[2]) + 'd' + str(e[3]) + ')'
    a1 = '(d' + str(d1_index[0]) + 'd' + str(d1_index[1]) + 'd' + str(d1_index[2]) + ')'
    a2 = '(d' + str(d2_index[0]) + 'd' + str(d2_index[1]) + ')'
    permutations_d[a0] = permutations_d[a1] * permutations_d[a2]

class colored_point:
    def __init__(self, coordinates, color, size):
        self.coordinates = coordinates
        self.color = color
        self.size = size
    def __str__(self):
        return (str(self.coordinates) + str(self.color) + str(self.size))
    
def permutate(_list, permutation):
    c = 1
    copy =[]
    copy2 = []
    for a in _list:
        copy.append(colored_point(a.coordinates, a.color, a.size))
        copy2.append(colored_point(a.coordinates, a.color, a.size))
    for i in permutation:  
        copy[c].color =  copy2[i].color
        c+=1
    return copy
permutations_n = {}
index = 1
for val in permutations_d.values():
    permutations_n[index] = val
    index += 1

c = 0
cube_s = cube(center=(0 + c, 0 + c, 0 + c), size=10, opacity ='0.1', frame_thickness=5, frame_color='black')
points = []
points_size = 200
points.append(colored_point((0,0,0), 'red', points_size))
points.append(colored_point((-5,-5,-5), 'blue',  points_size))
points.append(colored_point((-5,5,-5), 'red', points_size))
points.append(colored_point((5,5,-5), 'yellow',  points_size))
points.append(colored_point((5,-5,-5), 'red', points_size))
points.append(colored_point((-5,-5,5), 'red',  points_size))
points.append(colored_point((-5,5,5), 'green',  points_size))
points.append(colored_point((5,5,5), 'red', points_size))
points.append(colored_point((5,-5,5), 'red', points_size))
points_sum = point3d(points[1].coordinates, color=points[1].color, size=points[1].size)
```

```python
def drawColoredCube(xyz, n, points):
    cube_2 = cube(center=xyz, size=10, opacity ='0.1', frame_thickness=5, frame_color='black')
    points2 = permutate(points[:], permutations_n[n])
    for p in points2:
        p.coordinates = (p.coordinates[0] + xyz[0], p.coordinates[1] + xyz[1], p.coordinates[2] + xyz[2])
    points_sum2 = point3d(points2[1].coordinates, color=points2[1].color, size=points2[1].size)
    for i in range(1, 9):
        points_sum2 += point3d(points2[i].coordinates, color=points2[i].color, size=points[i].size)
        c = 0.5
        coordinatesnew = (points2[i].coordinates[0] + c, points2[i].coordinates[1] + c,points2[i].coordinates[2] + c)
        zaxis = line3d([xyz, (xyz[0], xyz[1], xyz[2] + 5)], thickness=15, color='blue', opacity=1, )
        yaxis = line3d([xyz, (xyz[0], xyz[1]  + 5, xyz[2])], thickness=15, color='red', opacity=1, )
        xaxis = line3d([xyz, (xyz[0]  + 5, xyz[1], xyz[2])], thickness=15, color='green', opacity=1, )
    return cube_2 + points_sum2 + zaxis + yaxis + xaxis

drawsum = drawColoredCube((0, 0, 0),  1, points)
drawsum += drawColoredCube((-15, 0, 0), 19,points)
drawsum += drawColoredCube((-30, 0, 0), 9,points)
drawsum += drawColoredCube((-45, 0, 0),  24,points)
show(drawsum, frame=False)
```
